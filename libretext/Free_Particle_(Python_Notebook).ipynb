```python
import matplotlib.pyplot as plt
import numpy as np

# Defining Psi and PsiC (complex conjugate) functions
def psi(x,k): return (1.0/np.sqrt(2.0*np.pi))*(np.cos(k*x)+np.sin(k*x)*1j)
def psiC(x,k): return (1.0/np.sqrt(2.0*np.pi))*(np.cos(k*x)-np.sin(k*x)*1j)

# Reading the input variables from the user
k = float(input("Enter the value for k (in Angstroms-1) = "))
xmax = float(input("Enter the maximum value for x (in Angstroms) = "))

# Generating the wavefunction graph
plt.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
x = np.linspace(-xmax, xmax, 900)
fig, ax = plt.subplots()
lim1=1/np.sqrt(2*np.pi) # Maximum value of the wavefunction
ax.axis([-xmax,xmax,-lim1*1.1,lim1*1.1]) # Defining the limits to be plot in the graph
ax.plot(x, psi(x,k).real, label="Real", color="blue") # Ploting the real part of the wavefunction
ax.plot(x, psi(x,k).imag, label="Imag.", color="red") # Ploting the real part of the wavefunction
ax.hlines(0.0, -xmax, xmax, linewidth=1.8, linestyle='--', color="black") # Adding a horizontal line at 0
# Now we define labels, legend, etc
ax.legend(loc=2);
ax.set_xlabel(r'$x$ (Angstroms)')
ax.set_ylabel(r'$\psi_k(x)$')
str1=r"$k$ = "+str(k)+r" A$^{-1}$"
plt.title('Wavefunction for '+str1)
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0.0)

# Generating the probability density graph
fig, ax = plt.subplots()
ax.axis([-xmax,xmax,0.0,lim1*lim1*1.4]) # Defining the limits to be plot in the graph
ax.plot(x, psi(x,k)*psiC(x,k),label="Probability Density", color="green") # Plotting the probability density
# Now we define labels, legend, etc
ax.legend(loc=2);
ax.set_xlabel(r'$x$ (Angstroms)')
ax.set_ylabel(r'$\left|\psi_k(x)\right|^2$')
str1=r"$k$ = "+str(k)+r" A$^{-1}$"
plt.title('Probability Density for '+str1)
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0.0)

# Show the plots on the screen once the code reaches this point
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import warnings

# Ignore common warnings
warnings.filterwarnings('ignore')

# --- Physical Constants ---
H_BAR = 1.054571817e-34  # Reduced Planck's constant in J*s
M_E = 9.1093837e-31       # Mass of an electron in kg

#======================================================================
# PART 1: CORE CALCULATION FUNCTIONS (Final Corrected Version)
#======================================================================

def gaussian_wave_packet(x, x0, k0, sigma_x):
    """
    Creates a normalized Gaussian wave packet centered at x0 with average
    wavenumber k0 and spatial width sigma_x.
    """
    norm = 1.0 / np.sqrt(sigma_x * np.sqrt(np.pi))
    gauss = np.exp(-(x - x0)**2 / (2.0 * sigma_x**2))
    plane_wave = np.exp(1j * k0 * x)
    return norm * gauss * plane_wave

def prob_density(psi):
    """Calculates the probability density from a wavefunction."""
    return (np.conjugate(psi) * psi).real

def evolve_wave_packet(psi_initial, x, t_sec, m_kg):
    """
    Evolves an initial Gaussian wave packet over time t.
    This uses the analytic solution for free particle evolution.
    """
    # Find initial width sigma_x0 from the initial probability density.
    # The `x` array is in Angstroms, so sigma_x0 will also be in Angstroms.
    p0 = prob_density(psi_initial)
    x0 = x[np.argmax(p0)]
    sigma_x0_A = np.sqrt(np.sum(p0 * (x - x0)**2) / np.sum(p0))

    # Convert initial width from Angstroms to meters for SI unit calculation.
    sigma_x0_m = sigma_x0_A * 1e-10

    # Calculate the time-dependent width in meters using SI units.
    sigma_t_m = sigma_x0_m * np.sqrt(1 + (H_BAR * t_sec / (m_kg * sigma_x0_m**2))**2)

    # Convert the new width back to Angstroms for creating the new Gaussian profile.
    sigma_t_A = sigma_t_m * 1e10

    # Evolved wavefunction (ignoring overall phase and group velocity for simplicity)
    norm = 1.0 / np.sqrt(sigma_t_A * np.sqrt(np.pi))
    gauss = np.exp(-(x - x0)**2 / (2.0 * sigma_t_A**2))

    return norm * gauss


#======================================================================
# PART 2: AUTOMATED SWEEP & PLOTTING FUNCTIONS
#======================================================================

def plot_uncertainty_principle_sweep(k0, dk_small, dk_large, x_range_A):
    """
    Illustrates the Heisenberg Uncertainty Principle by comparing a packet
    with small momentum uncertainty to one with large momentum uncertainty.
    """
    print("--- Generating Standard Derivation Sweep: Uncertainty Principle ---")
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    fig.suptitle('Standard Derivation: The Uncertainty Principle (Static Picture at t=0)', fontsize=16)

    x = np.linspace(-x_range_A / 2, x_range_A / 2, 1000)

    # Case 1: Small Momentum Uncertainty -> Large Position Uncertainty
    sigma_x_large = 1 / (2 * dk_small)
    psi1 = gaussian_wave_packet(x, 0, k0, sigma_x_large)
    prob1 = prob_density(psi1)

    ax1.plot(x, prob1, color='blue')
    ax1.fill_between(x, prob1, color='blue', alpha=0.2)
    ax1.set_title(f'Low Momentum Uncertainty ($\Delta k$ = {dk_small} Å⁻¹)')
    ax1.set_xlabel('Position (Angstroms)')
    ax1.set_ylabel('$|\\Psi(x)|^2$')
    ax1.grid(True, linestyle='--', alpha=0.6)

    # Case 2: Large Momentum Uncertainty -> Small Position Uncertainty
    sigma_x_small = 1 / (2 * dk_large)
    psi2 = gaussian_wave_packet(x, 0, k0, sigma_x_small)
    prob2 = prob_density(psi2)

    ax2.plot(x, prob2, color='magenta')
    ax2.fill_between(x, prob2, color='magenta', alpha=0.2)
    ax2.set_title(f'High Momentum Uncertainty ($\Delta k$ = {dk_large} Å⁻¹)')
    ax2.set_xlabel('Position (Angstroms)')
    ax2.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

def plot_dynamic_evolution_sweep(k0, dk, time_points_fs, m_ue):
    """
    Shows the time evolution (spreading) of a single Gaussian wave packet.
    *** THIS FUNCTION HAS BEEN CORRECTED TO DYNAMICALLY RESIZE THE PLOT WINDOW. ***
    """
    print("\n--- Generating Dynamic Perspective Sweep: Wave Packet Spreading ---")
    print(f"Evolving a wave packet over times: {time_points_fs} fs")

    # --- FIX: Dynamically determine the plot range ---
    # 1. Calculate the initial width and convert to SI units
    m_kg = m_ue * M_E
    sigma_x0_A = 1 / (2 * dk)
    sigma_x0_m = sigma_x0_A * 1e-10

    # 2. Find the width at the FINAL time step to set the plot boundary
    t_final_sec = time_points_fs[-1] * 1e-15
    sigma_t_final_m = sigma_x0_m * np.sqrt(1 + (H_BAR * t_final_sec / (m_kg * sigma_x0_m**2))**2)
    sigma_t_final_A = sigma_t_final_m * 1e10

    # 3. Set the x-axis range to be large enough to see the most spread-out packet
    # A plot width of +/- 5 standard deviations gives a good view.
    plot_width_A = sigma_t_final_A * 10

    # 4. Create the x-axis array based on this dynamic range
    x = np.linspace(-plot_width_A / 2, plot_width_A / 2, 2000) # Increased points for smooth wide curve

    # --- Plotting proceeds on the new, wider x-axis ---
    plt.figure(figsize=(10, 6))
    plt.title('Dynamic Perspective: Time Evolution of Probability Distribution')

    # Initial Packet at t=0
    psi_initial = gaussian_wave_packet(x, 0, k0, sigma_x0_A)
    prob_initial = prob_density(psi_initial)
    plt.plot(x, prob_initial, 'k--', label='Initial (t=0 fs)', lw=2)

    colors = plt.cm.viridis(np.linspace(0.3, 1, len(time_points_fs)-1))

    for i, t_fs in enumerate(time_points_fs):
        if t_fs == 0: continue
        t_sec = t_fs * 1e-15

        psi_evolved = evolve_wave_packet(psi_initial, x, t_sec, m_kg)
        prob_evolved = prob_density(psi_evolved)

        plt.plot(x, prob_evolved, color=colors[i-1], label=f't = {t_fs} fs')

    plt.xlabel('Position (Angstroms)')
    plt.ylabel('$|\\Psi(x,t)|^2$ (Probability Density)')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

#======================================================================
# MAIN EXECUTION BLOCK
#======================================================================

if __name__ == "__main__":
    print("Free Particle: Automated Dynamics Explorer (Corrected)")
    print("="*50)

    # --- Fixed Parameters for the Sweeps ---
    X_RANGE_STATIC_A = 40.0 # Fixed range for the static uncertainty plot
    K0 = 1.0                 # Average wavenumber in Angstroms^-1 (sets momentum)
    PARTICLE_MASS_UE = 1.0   # Mass in units of electron mass

    # --- Execute Standard Derivation Sweep ---
    plot_uncertainty_principle_sweep(
        k0=K0,
        dk_small=0.2,  # Corresponds to a wide packet
        dk_large=1.0,  # Corresponds to a narrow packet
        x_range_A=X_RANGE_STATIC_A
    )

    # --- Execute Dynamic (SQC-Inspired) Sweep ---
    time_points = [0.0, 5.0, 15.0, 30.0] # Femtoseconds
    plot_dynamic_evolution_sweep(
        k0=K0,
        dk=0.8, # A reasonably localized initial packet
        time_points_fs=time_points,
        m_ue=PARTICLE_MASS_UE
    )

    print("\n--- All sweeps complete. ---")
```
