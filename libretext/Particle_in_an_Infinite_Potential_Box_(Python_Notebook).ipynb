```python
import matplotlib.pyplot as plt
import numpy as np

# Defining the wavefunction
def psi(x,n,L): return np.sqrt(2.0/L)*np.sin(float(n)*np.pi*x/L)

# Reading the input variables from the user
n = int(input("Enter the value for the quantum number n = "))
L = float(input("Enter the size of the box in Angstroms = "))

# Generating the wavefunction graph
plt.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
x = np.linspace(0, L, 900)
fig, ax = plt.subplots()
lim1=np.sqrt(2.0/L) # Maximum value of the wavefunction
ax.axis([0.0,L,-1.1*lim1,1.1*lim1]) # Defining the limits to be plot in the graph
str1=r"$n = "+str(n)+r"$"
ax.plot(x, psi(x,n,L), linestyle='--', label=str1, color="orange", linewidth=2.8) # Plotting the wavefunction
ax.hlines(0.0, 0.0, L, linewidth=1.8, linestyle='--', color="black") # Adding a horizontal line at 0
# Now we define labels, legend, etc
ax.legend(loc=2);
ax.set_xlabel(r'$L$')
ax.set_ylabel(r'$\psi_n(x)$')
plt.title('Wavefunction')
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0.0)

# Generating the probability density graph
fig, ax = plt.subplots()
ax.axis([0.0,L,0.0,lim1*lim1*1.1])
str1=r"$n = "+str(n)+r"$"
ax.plot(x, psi(x,n,L)*psi(x,n,L), label=str1, linewidth=2.8)
ax.legend(loc=2);
ax.set_xlabel(r'$L$')
ax.set_ylabel(r'$|\psi_n|^2(x)$')
plt.title('Probability Density')
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0.0)

# Show the plots on the screen once the code reaches this point
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import warnings

# Ignore common warnings
warnings.filterwarnings('ignore')

# --- Physical Constants ---
H_BAR = 1.054571817e-34  # Reduced Planck's constant in J*s
M_E = 9.1093837e-31       # Mass of an electron in kg
EV_TO_JOULE = 1.60218e-19 # Conversion factor from eV to Joules

#======================================================================
# PART 1: STANDARD DERIVATION FUNCTIONS
# Inspired by the provided notebook
#======================================================================

def psi(x, n, L):
    """
    Calculates the normalized wavefunction for a particle in a 1D box.

    """
    return np.sqrt(2.0 / L) * np.sin(float(n) * np.pi * x / L)

def prob_density(x, n, L):
    """
    Calculates the probability density for a particle in a 1D box.
    """
    wavefunction = psi(x, n, L)
    return wavefunction**2

def energy(n, L, m):
    """
    Calculates the energy of the nth level for a particle in a 1D box.

    """
    # Using h-bar instead of h for consistency with quantum formalism
    h = H_BAR * 2 * np.pi
    prefactor = (h**2 / (m * 8)) * (1e10)**2 / EV_TO_JOULE # Prefactor to get eV with L in Angstroms
    return prefactor * ((float(n) / L)**2)

#======================================================================
# PART 2: STOCHASTIC-QUANTUM CORRESPONDENCE (SQC) FUNCTIONS
# Based on the principles from the Barandes papers
#======================================================================

def calculate_unitary_matrix(n_max, L_nm, m_kg, t_sec, N_points):
    """
    Calculates the unitary time-evolution matrix U(t) in the position basis.
    Each element U_ij is the quantum propagator from position j to i.

    Args:
        n_max (int): Number of energy levels to include in the summation.
        L_nm (float): Length of the box in nanometers.
        m_kg (float): Mass of the particle in kg.
        t_sec (float): Time of evolution in seconds.
        N_points (int): Number of discrete points to discretize the box.

    Returns:
        np.ndarray: The (N_points x N_points) unitary matrix U(t).
    """
    L_m = L_nm * 1e-9 # Convert length to meters
    x_positions = np.linspace(0, L_m, N_points)

    # Create meshgrid for vectorized calculation
    # xj corresponds to the starting position, xi to the final
    xi, xj = np.meshgrid(x_positions, x_positions)

    U = np.zeros((N_points, N_points), dtype=complex)

    # Sum over energy eigenstates (quantum numbers n)
    for n in range(1, n_max + 1):
        # Energy E_n in Joules
        E_n_joule = (n**2 * np.pi**2 * H_BAR**2) / (2 * m_kg * L_m**2)

        # Wavefunction product psi_n(xi) * psi_n(xj)
        psi_product = (2.0 / L_m) * np.sin(n * np.pi * xi / L_m) * np.sin(n * np.pi * xj / L_m)

        # Time evolution factor
        time_evolution_factor = np.exp(-1j * E_n_joule * t_sec / H_BAR)

        # Add the contribution of this energy level to the unitary matrix
        U += psi_product * time_evolution_factor

    # The factor for the discretization dx
    dx = L_m / (N_points -1) if N_points > 1 else L_m
    return U * dx


def calculate_stochastic_matrix(U):
    """
    Calculates the stochastic transition matrix Gamma(t) from the unitary matrix U(t).
    Gamma_ij = |U_ij|^2
    """
    return np.abs(U)**2

#======================================================================
# PART 3: PLOTTING AND VISUALIZATION
#======================================================================

def plot_standard_derivation(n, L_angstrom):
    """
    Plots the wavefunction and probability density for a given quantum state n.
    """
    print(f"\n--- Plotting Standard Derivation for n={n} ---")
    x = np.linspace(0, L_angstrom, 500)
    wavefunc = psi(x, n, L_angstrom)
    prob_dens = prob_density(x, n, L_angstrom)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    fig.suptitle('Standard Derivation: Stationary State Properties', fontsize=16)

    # Plot Wavefunction
    ax1.plot(x, wavefunc, 'r-', label=f'$\\psi_{n}(x)$')
    ax1.hlines(0.0, 0.0, L_angstrom, linewidth=1.0, linestyle='--', color="black")
    ax1.set_title(f'Wavefunction for n={n}')
    ax1.set_xlabel('Position (Angstroms)')
    ax1.set_ylabel('$\\psi_n(x)$')
    ax1.grid(True, linestyle='--', alpha=0.6)
    ax1.legend()

    # Plot Probability Density
    ax2.plot(x, prob_dens, 'b-', label=f'$|\\psi_{n}(x)|^2$')
    ax2.hlines(0.0, 0.0, L_angstrom, linewidth=1.0, linestyle='--', color="black")
    ax2.set_title(f'Probability Density for n={n}')
    ax2.set_xlabel('Position (Angstroms)')
    ax2.set_ylabel('$|\\psi_n(x)|^2$')
    ax2.fill_between(x, prob_dens, color='blue', alpha=0.2)
    ax2.grid(True, linestyle='--', alpha=0.6)
    ax2.legend()

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

def plot_sqc_derivation(gamma_matrix, p_initial, p_final, L_angstrom, t_femtoseconds):
    """
    Plots the results of the SQC derivation.
    """
    print(f"\n--- Plotting SQC Derivation for t={t_femtoseconds} fs ---")
    N_points = len(p_initial)
    x_positions = np.linspace(0, L_angstrom, N_points)

    fig = plt.figure(figsize=(14, 6))
    gs = fig.add_gridspec(2, 2, height_ratios=[1, 10])

    ax1 = fig.add_subplot(gs[1, 0])
    ax2 = fig.add_subplot(gs[1, 1])
    cax = fig.add_subplot(gs[0, 0])

    fig.suptitle('Stochastic-Quantum Correspondence: System Evolution', fontsize=16)

    # Plot Stochastic Matrix Gamma(t)
    im = ax1.imshow(gamma_matrix, origin='lower', aspect='auto',
                    extent=[0, L_angstrom, 0, L_angstrom], cmap='magma')
    ax1.set_title('Stochastic Matrix $\\Gamma(t)$ at t={:.1f} fs'.format(t_femtoseconds))
    ax1.set_xlabel('Starting Position $x_j$ (Angstroms)')
    ax1.set_ylabel('Final Position $x_i$ (Angstroms)')
    fig.colorbar(im, cax=cax, orientation='horizontal')
    cax.set_title("Transition Probability $\\Gamma_{ij}(t)$")


    # Plot Evolved Probability Distribution
    ax2.plot(x_positions, p_initial, 'g--', label='Initial Distribution (t=0)')
    ax2.plot(x_positions, p_final, 'b-', label=f'Evolved Distribution (t={t_femtoseconds} fs)')
    ax2.set_title('Evolution of Probability Distribution')
    ax2.set_xlabel('Position (Angstroms)')
    ax2.set_ylabel('Probability')
    ax2.grid(True, linestyle='--', alpha=0.6)
    ax2.legend()

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()


#======================================================================
# MAIN EXECUTION BLOCK
#======================================================================

if __name__ == "__main__":
    print("Particle in a Box: A Tale of Two Derivations")
    print("="*50)

    # --- User Inputs ---
    L_A = float(input("Enter the size of the box (L) in Angstroms (e.g., 10): "))
    n_state = int(input("Enter the quantum number (n) for the standard plot (e.g., 3): "))
    mass_ue = float(input("Enter particle mass in units of electron mass (e.g., 1.0): "))

    print("\nFor the SQC derivation, we will see how a particle localized at a point evolves.")
    initial_pos_A = float(input(f"Enter the initial position of the particle (0 to {L_A} Angstroms): "))
    t_fs = float(input("Enter a time for evolution (t) in femtoseconds (e.g., 5.0): "))

    # --- Constants and Parameters for SQC ---
    L_nm_sqc = L_A / 10.0 # Convert Angstrom to nm
    mass_kg_sqc = mass_ue * M_E
    t_sec_sqc = t_fs * 1e-15
    N_POINTS_SQC = 101 # Number of discrete points in the box
    N_LEVELS_SQC = 50  # Number of energy levels to sum over for accuracy

    # --- Run Standard Derivation ---
    plot_standard_derivation(n_state, L_A)

    # --- Run SQC Derivation ---
    print("\nCalculating SQC matrices... (This may take a moment)")
    # 1. Calculate the unitary evolution matrix U(t)
    unitary_matrix = calculate_unitary_matrix(
        n_max=N_LEVELS_SQC,
        L_nm=L_nm_sqc,
        m_kg=mass_kg_sqc,
        t_sec=t_sec_sqc,
        N_points=N_POINTS_SQC
    )

    # 2. Calculate the stochastic transition matrix Gamma(t)
    stochastic_matrix = calculate_stochastic_matrix(unitary_matrix)

    # 3. Define an initial probability distribution (particle localized at one point)
    p_initial = np.zeros(N_POINTS_SQC)
    # Find the closest grid point to the user's input
    start_index = np.abs(np.linspace(0, L_A, N_POINTS_SQC) - initial_pos_A).argmin()
    p_initial[start_index] = 1.0 # Probability is 1 at the starting index

    # 4. Evolve the system using the stochastic matrix: p(t) = Gamma(t) * p(0)
    # Note: In matrix form this is a multiplication.
    # Since Gamma_ij is p(i|j), p_i(t) = sum_j Gamma_ij(t) * p_j(0)
    p_final = stochastic_matrix @ p_initial

    # 5. Plot the results
    plot_sqc_derivation(stochastic_matrix, p_initial, p_final, L_A, t_fs)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import warnings

# Ignore common warnings
warnings.filterwarnings('ignore')

# --- Physical Constants ---
H_BAR = 1.054571817e-34  # Reduced Planck's constant in J*s
M_E = 9.1093837e-31       # Mass of an electron in kg
EV_TO_JOULE = 1.60218e-19 # Conversion factor from eV to Joules
K_B = 1.380649e-23        # Boltzmann constant in J/K

#======================================================================
# PART 1: CORE CALCULATION FUNCTIONS (Unchanged)
#======================================================================

def psi(x, n, L):
    """Calculates the normalized wavefunction."""
    return np.sqrt(2.0 / L) * np.sin(float(n) * np.pi * x / L)

def prob_density(x, n, L):
    """Calculates the probability density."""
    return psi(x, n, L)**2

def energy(n, L_angstrom, m_kg):
    """Calculates the energy of the nth level in Joules."""
    L_m = L_angstrom * 1e-10
    return (n**2 * np.pi**2 * H_BAR**2) / (2 * m_kg * L_m**2)

def calculate_unitary_matrix(n_max, L_nm, m_kg, t_sec, N_points):
    """Calculates the unitary time-evolution matrix U(t)."""
    L_m = L_nm * 1e-9
    x_positions = np.linspace(0, L_m, N_points)
    xi, xj = np.meshgrid(x_positions, x_positions)
    U = np.zeros((N_points, N_points), dtype=complex)

    for n in range(1, n_max + 1):
        # Use energy in Joules for the evolution formula
        E_n_joule = energy(n, L_nm * 10, m_kg)
        psi_product = (2.0 / L_m) * np.sin(n * np.pi * xi / L_m) * np.sin(n * np.pi * xj / L_m)
        time_evolution_factor = np.exp(-1j * E_n_joule * t_sec / H_BAR)
        U += psi_product * time_evolution_factor

    dx = L_m / (N_points - 1) if N_points > 1 else L_m
    return U * dx

def calculate_stochastic_matrix(U):
    """Calculates the stochastic transition matrix Gamma(t)."""
    return np.abs(U)**2

#======================================================================
# PART 2: PLOTTING FUNCTIONS (Mostly Unchanged)
#======================================================================

# (plot_standard_sweep, plot_sqc_time_sweep, plot_sqc_position_sweep are omitted for brevity
# but would be included in a full script)


#======================================================================
# PART 3: NEW STATISTICAL MECHANICS CONNECTION
#======================================================================

def calculate_thermal_prob_density(x_A, T, n_max, L_A, m_kg):
    """
    Calculates the thermal equilibrium probability density at temperature T.
    """
    x_m = x_A * 1e-10
    L_m = L_A * 1e-10

    total_prob = np.zeros_like(x_m)
    partition_func = 0.0

    # Sum over states weighted by the Boltzmann factor
    for n in range(1, n_max + 1):
        E_n = energy(n, L_A, m_kg)
        boltzmann_factor = np.exp(-E_n / (K_B * T))

        partition_func += boltzmann_factor

        # |psi_n(x)|^2
        pd_n = (2.0 / L_m) * (np.sin(n * np.pi * x_m / L_m))**2
        total_prob += pd_n * boltzmann_factor

    # Normalize by the partition function
    return total_prob / partition_func

def plot_thermal_equilibrium_sweep(T_values, n_max, L_A, m_kg):
    """
    Plots the thermal probability distribution for several temperatures.
    """
    print("\n--- Generating Statistical Mechanics Connection Sweep ---")
    plt.figure(figsize=(10, 6))
    plt.title('Thermal Equilibrium Positional Probability')

    x = np.linspace(0, L_A, 500)
    colors = plt.cm.plasma(np.linspace(0, 1, len(T_values)))

    for i, T in enumerate(T_values):
        thermal_pd = calculate_thermal_prob_density(x, T, n_max, L_A, m_kg)
        # Normalize area to 1 for plotting probability density
        thermal_pd /= np.trapz(thermal_pd, x)
        plt.plot(x, thermal_pd, label=f'T = {T} K', color=colors[i])

    # For comparison, the classical high-temperature limit is a uniform distribution
    plt.axhline(1.0 / L_A, color='gray', linestyle=':', label='Classical Limit (High T)')

    plt.xlabel('Position (Angstroms)')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()


#======================================================================
# MAIN EXECUTION BLOCK
#======================================================================

if __name__ == "__main__":
    print("Connecting SQC for Particle in a Box to Statistical Mechanics")
    print("="*60)

    # --- Fixed Parameters ---
    BOX_LENGTH_A = 10.0  # Angstroms
    PARTICLE_MASS_UE = 1.0 # In units of electron mass
    m_kg = PARTICLE_MASS_UE * M_E

    # --- Execute Statistical Mechanics Sweep ---
    # Show the probability distribution for different temperatures.
    # We choose temperatures that highlight different physical regimes.
    # T=1K is very low, T=300K is room temp, T=5000K is very high.
    temperatures = [1, 300, 5000, 50000]

    plot_thermal_equilibrium_sweep(
        T_values=temperatures,
        n_max=50, # Sum over enough states for convergence at high T
        L_A=BOX_LENGTH_A,
        m_kg=m_kg
    )
```

```python
import matplotlib.pyplot as plt
import numpy as np
import warnings

# Ignore common warnings
warnings.filterwarnings('ignore')

# --- Physical Constants ---
H_BAR = 1.054571817e-34  # Reduced Planck's constant in J*s
M_E = 9.1093837e-31       # Mass of an electron in kg
EV_TO_JOULE = 1.60218e-19 # Conversion factor from eV to Joules
K_B = 1.380649e-23        # Boltzmann constant in J/K

#======================================================================
# PART 1: CORE CALCULATION FUNCTIONS
#======================================================================

def energy(n, L_angstrom, m_kg):
    """Calculates the energy of the nth level in Joules."""
    L_m = L_angstrom * 1e-10
    return (n**2 * np.pi**2 * H_BAR**2) / (2 * m_kg * L_m**2)

#======================================================================
# PART 2: NEW THERMODYNAMICS FUNCTIONS
#======================================================================

def calculate_thermodynamic_properties(T, n_max, L_A, m_kg):
    """
    Calculates Z, U, F, and S for a given temperature T.
    """
    # Pre-calculate energies for all n
    n_vals = np.arange(1, n_max + 1)
    E_n_vals = energy(n_vals, L_A, m_kg)

    # Calculate Boltzmann factors
    beta = 1.0 / (K_B * T)
    boltzmann_factors = np.exp(-beta * E_n_vals)

    # 1. Partition Function (Z)
    Z = np.sum(boltzmann_factors)

    # 2. Average Internal Energy (U)
    U = np.sum(E_n_vals * boltzmann_factors) / Z

    # 3. Helmholtz Free Energy (F)
    F = -K_B * T * np.log(Z)

    # 4. Entropy (S)
    S = (U - F) / T

    return Z, U, F, S

def plot_thermodynamics_sweep(T_range, n_max, L_A, m_kg):
    """
    Calculates and plots the key thermodynamic properties over a range of temperatures.
    """
    print("\n--- Generating Thermodynamics Sweep ---")

    # Calculate properties for each temperature in the range
    U_vals = []
    S_vals = []
    F_vals = []

    for T in T_range:
        if T == 0: # Avoid division by zero
            Z, U, F, S = calculate_thermodynamic_properties(1e-9, n_max, L_A, m_kg)
        else:
            Z, U, F, S = calculate_thermodynamic_properties(T, n_max, L_A, m_kg)
        U_vals.append(U)
        S_vals.append(S)
        F_vals.append(F)

    U_vals = np.array(U_vals)
    S_vals = np.array(S_vals)
    F_vals = np.array(F_vals)

    # 5. Heat Capacity (Cv) - calculated as the derivative of U w.r.t T
    # We divide by K_B to get a dimensionless value
    Cv_vals = np.gradient(U_vals, T_range) / K_B

    # --- Plotting ---
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    fig.suptitle('Thermodynamic Properties of a Quantum Particle in a Box', fontsize=16)

    # Plot 1: Average Energy vs. Temperature
    ax1 = axes[0, 0]
    # Plot energy in units of the ground state energy E1
    E1 = energy(1, L_A, m_kg)
    ax1.plot(T_range, U_vals / E1, color='dodgerblue')
    ax1.set_title('Average Internal Energy (U)')
    ax1.set_xlabel('Temperature (K)')
    ax1.set_ylabel('Energy ($U/E_1$)')

    # Plot 2: Heat Capacity vs. Temperature
    ax2 = axes[0, 1]
    ax2.plot(T_range, Cv_vals, color='orangered')
    ax2.set_title('Heat Capacity ($C_V$)')
    ax2.set_xlabel('Temperature (K)')
    ax2.set_ylabel('Heat Capacity ($C_V / k_B$)')
    # Classical limit for 1D translation
    ax2.axhline(0.5, color='gray', linestyle=':', label='Classical Limit (k_B/2)')
    ax2.legend()

    # Plot 3: Entropy vs. Temperature
    ax3 = axes[1, 0]
    ax3.plot(T_range, S_vals / K_B, color='forestgreen')
    ax3.set_title('Entropy (S)')
    ax3.set_xlabel('Temperature (K)')
    ax3.set_ylabel('Entropy ($S / k_B$)')

    # Plot 4: Helmholtz Free Energy vs. Temperature
    ax4 = axes[1, 1]
    ax4.plot(T_range, F_vals / E1, color='purple')
    ax4.set_title('Helmholtz Free Energy (F)')
    ax4.set_xlabel('Temperature (K)')
    ax4.set_ylabel('Free Energy ($F/E_1$)')

    for ax in axes.flat:
        ax.grid(True, linestyle=':', alpha=0.7)

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

#======================================================================
# MAIN EXECUTION BLOCK
#======================================================================

if __name__ == "__main__":
    print("Connecting Statistical Mechanics to Thermodynamics")
    print("="*50)

    # --- Fixed Parameters ---
    BOX_LENGTH_A = 10.0      # Angstroms
    PARTICLE_MASS_UE = 1.0   # In units of electron mass
    m_kg = PARTICLE_MASS_UE * M_E

    # --- Execute Thermodynamics Sweep ---
    # Use a logarithmic scale for temperature to see low-T behavior clearly
    temperature_range = np.logspace(1, 5, 500) # From 10 K to 100,000 K

    plot_thermodynamics_sweep(
        T_range=temperature_range,
        n_max=50,
        L_A=BOX_LENGTH_A,
        m_kg=m_kg
    )
```
