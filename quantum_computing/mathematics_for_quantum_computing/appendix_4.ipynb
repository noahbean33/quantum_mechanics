# Appendix 4: Probability

The study of gambling, specifically the throwing of dice, led to the mathematical field of probability. Probability is the study of how likely an event is to occur given a number of possible outcomes. A real number between 0 and 1 is assigned to each event where 0 signifies the event has no chance of happening and 1 signifies the event will always happen. You can also multiply these numbers by 100 to get a percentage. All the probabilities for all possible outcomes must sum to 1.

In this chapter, we are going to cover the following main topics:

- Definitions
- Random variables

## Definitions

**Experiment:** The execution of a procedure that produces a random outcome. Examples: flipping a coin, rolling dice, measuring a qubit.

**Sample space (Ω):** The set of all possible outcomes of an experiment.
- Fair coin: Ω = {Heads, Tails}
- One die: Ω = {1, 2, 3, 4, 5, 6}
- Qubit measured in Z basis: Ω = {|0⟩, |1⟩}

**Event (E):** A subset of Ω. Every outcome is a subset of size 1. The set of all events is called an **event space** (ℱ). For a fair coin:
- ∅
- {Heads}
- {Tails}
- {Heads, Tails} = Ω

**Probability function (P):** Maps each event to a real number between 0 and 1:

$$P: \mathcal{F} \to [0, 1]$$

Properties:
- $P(\Omega) = 1$
- $P(E) \geq 0$ for all events E
- For mutually exclusive events: $P(E_1 \cup E_2) = P(E_1) + P(E_2)$

## Random Variables

An important concept in probability. Oftentimes, we are not interested in the actual outcome but some function of the outcome. Random variables are functions from the sample space Ω to a measurable space (usually the real numbers).

### Discrete random variables

Most random variables in quantum computing are discrete. Discrete random variables have distinct values, and their sample spaces are finite or countably infinite.

The **Probability Mass Function (PMF)** is defined as:

$$p_X(z) = P(X = z)$$

Properties:
- $p_X(z) \geq 0$ for all z
- $\sum_z p_X(z) = 1$

> *Figure 13.1 – The PMF of S, the sum of two dice rolled*

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Probability: PMF, Expected Value, Variance ---

# Single die roll
outcomes = np.array([1, 2, 3, 4, 5, 6])
pmf_die = np.ones(6) / 6  # uniform distribution

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# PMF of a single die
ax = axes[0]
ax.bar(outcomes, pmf_die, color='steelblue', edgecolor='black', alpha=0.8)
ax.set_xlabel('Outcome'); ax.set_ylabel('P(X = z)')
ax.set_title('PMF of a Single Die Roll')
ax.set_xticks(outcomes); ax.set_ylim(0, 0.25)
ax.axhline(y=1/6, color='red', linestyle='--', alpha=0.5, label=f'P = 1/6 ≈ {1/6:.4f}')
ax.legend()

# PMF of sum of two dice
sums = np.arange(2, 13)
pmf_two_dice = np.zeros(11)
for i in range(1, 7):
    for j in range(1, 7):
        pmf_two_dice[i + j - 2] += 1
pmf_two_dice /= 36

ax = axes[1]
ax.bar(sums, pmf_two_dice, color='coral', edgecolor='black', alpha=0.8)
ax.set_xlabel('Sum S'); ax.set_ylabel('P(S = z)')
ax.set_title('PMF of Sum of Two Dice')
ax.set_xticks(sums)
for s, p in zip(sums, pmf_two_dice):
    if p > 0.05:
        ax.text(s, p + 0.005, f'{p:.3f}', ha='center', fontsize=7)

# Simulated die rolls — histogram converging to PMF
np.random.seed(42)
n_rolls = 10000
rolls = np.random.randint(1, 7, size=n_rolls)
ax = axes[2]
ax.hist(rolls, bins=np.arange(0.5, 7.5, 1), density=True, color='mediumseagreen',
        edgecolor='black', alpha=0.8, label=f'{n_rolls} simulated rolls')
ax.axhline(y=1/6, color='red', linestyle='--', linewidth=2, label='True PMF = 1/6')
ax.set_xlabel('Outcome'); ax.set_ylabel('Relative Frequency')
ax.set_title(f'Simulation: {n_rolls} Die Rolls')
ax.set_xticks(outcomes); ax.legend()

plt.tight_layout()
plt.show()

# Expected value, variance, standard deviation
E_X = np.sum(outcomes * pmf_die)
E_X2 = np.sum(outcomes**2 * pmf_die)
Var_X = E_X2 - E_X**2
Std_X = np.sqrt(Var_X)

print("=== Single Die Roll Statistics ===")
print(f"E[X]   = Σ z·P(z) = {E_X:.4f}")
print(f"E[X²]  = Σ z²·P(z) = {E_X2:.4f}")
print(f"Var(X) = E[X²] - (E[X])² = {E_X2:.4f} - {E_X:.4f}² = {Var_X:.4f}")
print(f"σ      = √Var(X) = {Std_X:.4f}")
print(f"\nVerify with numpy:  mean={np.mean(outcomes):.4f}, var={np.var(outcomes):.4f}, std={np.std(outcomes):.4f}")

# Qubit measurement example
print("\n=== Qubit Measurement Probabilities ===")
theta = np.pi / 3
alpha = np.cos(theta / 2)
beta = np.sin(theta / 2)
p0 = abs(alpha)**2
p1 = abs(beta)**2
E_qubit = 0 * p0 + 1 * p1
print(f"|ψ⟩ = cos(π/6)|0⟩ + sin(π/6)|1⟩")
print(f"P(|0⟩) = |α|² = {p0:.4f}")
print(f"P(|1⟩) = |β|² = {p1:.4f}")
print(f"P(|0⟩) + P(|1⟩) = {p0 + p1:.4f}  (= 1 ✓)")
print(f"E[measurement] = {E_qubit:.4f}")
```

### The measures of a random variable

#### Expected value

The expected value of a random variable X:

$$E[X] = \sum_z z \cdot p_X(z)$$

For the roll of a single die:

$$E[X] = 1 \cdot \frac{1}{6} + 2 \cdot \frac{1}{6} + 3 \cdot \frac{1}{6} + 4 \cdot \frac{1}{6} + 5 \cdot \frac{1}{6} + 6 \cdot \frac{1}{6} = 3.5$$

The expected value is not always an actual possible value of the PMF.

#### Variance

The variance measures the spread of possible PMF values:

$$\text{Var}(X) = E[X^2] - (E[X])^2$$

For the roll of one die:

$$E[X^2] = 1^2 \cdot \frac{1}{6} + 2^2 \cdot \frac{1}{6} + \cdots + 6^2 \cdot \frac{1}{6} = \frac{91}{6} \approx 15.17$$

$$\text{Var}(X) = 15.17 - 3.5^2 = 2.92$$

#### Standard deviation

The standard deviation is the square root of the variance:

$$\sigma = \sqrt{\text{Var}(X)}$$

These are the three most important measures of a random variable: the expected value, the variance, and the standard deviation.
