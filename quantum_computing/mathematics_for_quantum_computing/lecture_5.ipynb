## Linearity

What makes a transform linear? This question gets to the heart of linear algebra. The concept of linearity ties together all the other concepts we have considered so far and the ones to come. Indeed, quantum mechanics is a linear theory. That's what makes linear algebra crucial to understanding quantum computing.

Real-life examples of non-linearity abound. For example, exercising 1 hour a day for 24 days does not give the same result as exercising 24 hours in 1 day. Watering a plant is another good non-linear example.

Linear relationships, on the other hand, are proportional. Speed is a good example. Exchange rates for money are also a good example.

> *Figure 5.1 – A graph of the dollars to euros function*

Crucially though, linearity requires functions to return 0 if 0 is the input to the function.

Let's take the simplest example, y = x:

> *Figure 5.2 – The line y = x*

**Homogeneity:** The property that scaling the input scales the output proportionally:

$$\alpha f(x) = f(\alpha x)$$

> *Table 5.1 – αf(x) and f(αx) at different values of x and α*

**Additivity:** The property that adding inputs adds outputs:

$$f(x) + f(z) = f(x + z)$$

> *Table 5.2 – f(x), f(z), and f(x+z) at different values of x and z*

These two properties, additivity and homogeneity, define **linearity**. To be a linear transformation, a transformation must have linearity.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Linearity test: f(x) = x vs f(x) = x + 1 ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

x = np.linspace(-3, 3, 100)

# Linear: f(x) = x
ax = axes[0]
ax.plot(x, x, 'b-', linewidth=2, label='f(x) = x')
ax.plot(0, 0, 'ro', markersize=8, label='f(0) = 0 ✓')
ax.set_title('LINEAR: f(x) = x\nHomogeneity ✓  Additivity ✓')
ax.legend(); ax.grid(True, alpha=0.3); ax.set_aspect('equal')
ax.set_xlim(-3, 3); ax.set_ylim(-3, 3)
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# Non-linear: f(x) = x + 1
ax = axes[1]
ax.plot(x, x + 1, 'r-', linewidth=2, label='f(x) = x + 1')
ax.plot(0, 1, 'ro', markersize=8, label='f(0) = 1 ≠ 0 ✗')
ax.set_title('NOT LINEAR: f(x) = x + 1\nf(0) ≠ 0')
ax.legend(); ax.grid(True, alpha=0.3); ax.set_aspect('equal')
ax.set_xlim(-3, 3); ax.set_ylim(-3, 3)
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()

# Verify linearity properties for f(x) = 2x
def f(v): return 2 * v

u = np.array([3, -1])
v = np.array([1, 4])
alpha = 5.0

print("Testing f(v) = 2v for linearity:")
print(f"  Additivity:   f(u+v) = {f(u+v)},  f(u)+f(v) = {f(u)+f(v)},  equal? {np.array_equal(f(u+v), f(u)+f(v))}")
print(f"  Homogeneity:  f(α·u) = {f(alpha*u)},  α·f(u) = {alpha*f(u)},  equal? {np.array_equal(f(alpha*u), alpha*f(u))}")
print(f"  f(0) = {f(np.array([0,0]))}  ✓")
```

## What is a Linear Transformation?

A linear transformation is a function T from a vector space U to a vector space V:

$$T: U \to V$$

To be linear, the transformation must preserve the operations of vector addition and scalar multiplication:

$$T(|u\rangle + |v\rangle) = T(|u\rangle) + T(|v\rangle)$$
$$T(s|u\rangle) = sT(|u\rangle)$$

It follows from these axioms that for any linear transformation T, T(**0**) has to equal the zero vector **0**.

### Describing linear transformations

There are many ways to describe a linear transformation.

#### A geometric description

Let's use reflection. Reflection is a very easy and intuitive transformation.

> *Figure 5.3 – A graphical depiction of the reflection transformation*

We can verify it passes both homogeneity and additivity tests:

> *Figure 5.4 – A test reflection for homogeneity*

> *Figure 5.6 – A test for additivity*

Therefore, our reflection transformation is a linear transformation.

#### An algebraic description

If you are dealing with n-tuples of numbers in ℝⁿ, you can say explicitly what the transform does:

$$T: \mathbb{R}^2 \to \mathbb{R}^3$$

The preceding equations fully define the linear transformation for every vector in our domain.

#### A basis vectors description

Since any vector in a vector space can be expressed as a linear combination of a set of basis vectors, if you describe what the transform does to a set of basis vectors, you've described the complete transformation.

## Representing Linear Transformations with Matrices

Through the magic of matrix-vector multiplication, a matrix is all you need to describe a linear transformation.

The dimension of the domain is the number of columns of the matrix, the dimension of the codomain is the number of rows of the matrix, and the actual transformation is the matrix itself. That is the power of a matrix!

### Matrices depend on the bases chosen

Our matrix depends on the basis we choose to represent it.

> *Figure 5.7 – A reflection transformation*

> *Figure 5.8 – A reflection transformation with basis vectors*

If we picked a different set of basis vectors, the matrix would be different as well. The matrix based on the canonical computational basis is called the **standard matrix** of the linear transformation. Matrices that represent the same linear transformation are called **similar**.

### Matrix multiplication and multiple transformations

Not only can matrices represent transformations, but they can also represent multiple transformations through matrix multiplication!

If we do our reflection transformation twice, we should get back the same vector:

$$A \cdot A = I$$

### The commutator

Matrices do not, in general, commute. Physicists use the **commutator** to represent "how much" two transformations or matrices commute:

$$[A, B] = AB - BA$$

If the commutator is zero for two transformations, then they commute. If it is non-zero, the operators are said to be incompatible. In quantum mechanics, observables such as momentum are represented by linear transformations. This leads to the famous **uncertainty principle** that states that two observables that do not commute cannot be measured simultaneously.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Reflection, Rotation, and Projection transformations ---
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Generate a set of test vectors (a square shape)
square = np.array([[1, 1], [1, 2], [2, 2], [2, 1], [1, 1]]).T

# 1) Reflection over x-axis
Reflect = np.array([[1, 0], [0, -1]])
reflected = Reflect @ square
ax = axes[0]
ax.plot(square[0], square[1], 'b-o', linewidth=2, label='Original')
ax.plot(reflected[0], reflected[1], 'r--o', linewidth=2, label='Reflected (x-axis)')
ax.axhline(y=0, color='k', linewidth=1); ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_xlim(-0.5, 3); ax.set_ylim(-3, 3)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend()
ax.set_title(f'Reflection over x-axis\nA = {Reflect.tolist()}')

# 2) Rotation by 45 degrees
theta = np.pi / 4
Rotate = np.array([[np.cos(theta), -np.sin(theta)],
                    [np.sin(theta),  np.cos(theta)]])
rotated = Rotate @ square
ax = axes[1]
ax.plot(square[0], square[1], 'b-o', linewidth=2, label='Original')
ax.plot(rotated[0], rotated[1], 'g--o', linewidth=2, label='Rotated 45°')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_xlim(-2, 3); ax.set_ylim(-1, 3.5)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend()
ax.set_title(f'Rotation by θ = 45°\nR(θ) = [[cos,−sin],[sin,cos]]')

# 3) Projection onto x-axis
Project = np.array([[1, 0], [0, 0]])
projected = Project @ square
ax = axes[2]
ax.plot(square[0], square[1], 'b-o', linewidth=2, label='Original')
ax.plot(projected[0], projected[1], 'm--o', linewidth=2, label='Projected (x-axis)')
for i in range(square.shape[1]):
    ax.plot([square[0, i], projected[0, i]], [square[1, i], projected[1, i]], 'k:', alpha=0.4)
ax.axhline(y=0, color='k', linewidth=1); ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_xlim(-0.5, 3); ax.set_ylim(-1, 3)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend()
ax.set_title(f'Projection onto x-axis\nP² = P? {np.array_equal(Project @ Project, Project)}')

plt.tight_layout()
plt.show()

# Verify properties
print("Reflection²  = I?", np.allclose(Reflect @ Reflect, np.eye(2)))
print("Projection²  = P?", np.allclose(Project @ Project, Project))

# Commutator
AB = Reflect @ Rotate
BA = Rotate @ Reflect
commutator = AB - BA
print(f"\n[Reflect, Rotate] = AB - BA =\n{np.round(commutator, 4)}")
print(f"Do they commute? {np.allclose(commutator, 0)}")
```

## Transformations Inspired by Euclid

In linear algebra, there are many "special" types of linear transformations that have names that connote concepts we have in our real world.

### Translation

Translation transforms all vectors in a vector space by a displacement vector:

$$T(|x\rangle) = |x\rangle + |d\rangle$$

> *Figure 5.10 – A graphical depiction of translation*

What's interesting about this type of translation is that it turns out to be **non-linear!**

$$T(|x\rangle + |y\rangle) \neq T(|x\rangle) + T(|y\rangle)$$

Also, $T(\mathbf{0}) \neq \mathbf{0}$.

### Rotation

Given a vector |v⟩ and an angle θ, R(θ) will rotate |v⟩ through an angle θ with respect to the x axis about the origin.

> *Figure 5.12 – A graphical depiction of a rotation transformation*

The rotation transformation for the computational basis vectors |0⟩ and |1⟩:

$$R(\theta)|0\rangle = \begin{bmatrix} \cos\theta \\ \sin\theta \end{bmatrix}, \quad R(\theta)|1\rangle = \begin{bmatrix} -\sin\theta \\ \cos\theta \end{bmatrix}$$

There is a theorem: if I give you the results of a transformation according to the computational basis, I can compute the standard matrix:

$$A = \begin{bmatrix} T(|0\rangle) & T(|1\rangle) \end{bmatrix}$$

The rotation matrix:

$$R(\theta) = \begin{bmatrix} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{bmatrix}$$

### Projection

Projection is another linear transformation used heavily in the measurement of qubits. If you have a linear transformation P, then if the following condition holds, it is a projection:

$$P^2 = P$$

## Linear Operators

Linear operators are linear transformations that map vectors from and to the **same** vector space. Indeed, reflections, rotations, and projections are all linear operators. In quantum, we put a "hat" or caret on the top of the letter of the linear operator:

$$\hat{R}: \mathbb{R}^2 \to \mathbb{R}^2$$

All linear operators are represented by **square matrices**. This leads to all types of special properties such as determinants, eigenvalues, and invertibility.

## Linear Functionals

A linear functional is a special case of a linear transformation that takes in a vector and spits out a scalar:

$$f: V \to F$$

The set of all linear functionals that can be defined on a vector space actually form their own vector space called the **dual vector space**. This concept is important to fully define a bra in bra-ket notation.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Change of Basis ---
# Standard (computational) basis
e1 = np.array([1, 0])
e2 = np.array([0, 1])

# Alternative basis
f1 = np.array([1, 1])
f2 = np.array([1, -1])

# A vector in the standard basis
v_standard = np.array([3, 1])

# Change of basis matrix: columns are the new basis vectors expressed in the old basis
B = np.column_stack([f1, f2])
# To get coords in basis F: solve B @ coords_F = v_standard
coords_F = np.linalg.solve(B, v_standard)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Standard basis view
ax = axes[0]
ax.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='blue', linewidth=2, label='|0⟩')
ax.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='red', linewidth=2, label='|1⟩')
ax.quiver(0, 0, v_standard[0], v_standard[1], angles='xy', scale_units='xy', scale=1, color='black', linewidth=2.5, label=f'v = {v_standard}')
ax.set_xlim(-1, 4); ax.set_ylim(-2, 3)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend()
ax.set_title(f'Standard Basis\nCoords: ({v_standard[0]}, {v_standard[1]})')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# Alternative basis view
ax = axes[1]
ax.quiver(0, 0, f1[0], f1[1], angles='xy', scale_units='xy', scale=1, color='green', linewidth=2, label='f₁ = [1,1]')
ax.quiver(0, 0, f2[0], f2[1], angles='xy', scale_units='xy', scale=1, color='orange', linewidth=2, label='f₂ = [1,-1]')
ax.quiver(0, 0, v_standard[0], v_standard[1], angles='xy', scale_units='xy', scale=1, color='black', linewidth=2.5, label=f'v (same vector!)')
# Show decomposition
ax.quiver(0, 0, coords_F[0]*f1[0], coords_F[0]*f1[1], angles='xy', scale_units='xy', scale=1, color='green', alpha=0.4, linewidth=1.5)
ax.quiver(coords_F[0]*f1[0], coords_F[0]*f1[1], coords_F[1]*f2[0], coords_F[1]*f2[1], angles='xy', scale_units='xy', scale=1, color='orange', alpha=0.4, linewidth=1.5)
ax.set_xlim(-1, 4); ax.set_ylim(-2, 3)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend(fontsize=9)
ax.set_title(f'Alternative Basis F\nCoords: ({coords_F[0]}, {coords_F[1]})')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()

print(f"Change of basis matrix B =\n{B}\n")
print(f"v in standard basis: {v_standard}")
print(f"v in basis F:        {coords_F}")
print(f"Verify: {coords_F[0]}*f₁ + {coords_F[1]}*f₂ = {coords_F[0]*f1 + coords_F[1]*f2}")
```

## A Change of Basis

We learned in Chapter 4, Vector Spaces, that a vector can have different coordinates depending on the basis that was chosen, but we didn't tell you how to go back and forth between bases.

We want to come up with a matrix B for a change of basis that takes us from one basis to another:

$$|v\rangle_F = B|v\rangle_C$$

This matrix B will convert the coordinates of a vector according to a basis C to the coordinates for the vector in the basis F.

The general formula for the change of basis matrix:

$$B = \begin{bmatrix} |0\rangle_F & |1\rangle_F \end{bmatrix}$$

This is basically saying that you have to express the basis vectors in the input basis as coordinates in the output basis.

Due to this, a change of basis is a linear transformation. With this change of basis matrix, we can change any vector expressed as coordinates in C to coordinates in F.
