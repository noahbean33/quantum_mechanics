# Appendix 2: Sigma Notation

Many quantum computing books will introduce sigma notation without ever really explaining it, and you will also see shorthand for it as well. I hope to demystify this notation when you encounter it in this book and others.

## Sigma

Sigma is the 18th letter of the Greek alphabet, and we are talking about the capital version, Σ. It signifies a summation:

$$\sum_{i=1}^{n} a_i = a_1 + a_2 + \cdots + a_n$$

More generally:

$$\sum_{i=m}^{n} a_i = a_m + a_{m+1} + \cdots + a_n$$

> *Figure 11.1 – A diagram of sigma notation*

Here is an easy example:

$$\sum_{i=1}^{4} i = 1 + 2 + 3 + 4 = 10$$

## Variations

You will see shorthand in a lot of quantum computing books, but don't let it throw you. The following are all equivalent:

$$\sum_{i=1}^{n} a_i = \sum_i a_i = \sum a_i$$

## Summation Rules

There are many properties of summations, but the two most important ones are:

$$\sum_{i=1}^{n} c \cdot a_i = c \sum_{i=1}^{n} a_i$$

where c is a constant, and:

$$\sum_{i=1}^{n} (a_i + b_i) = \sum_{i=1}^{n} a_i + \sum_{i=1}^{n} b_i$$

These are both based on the distributive property.

```python
import numpy as np

# --- Sigma Notation in Python ---

# Basic summation: Σ_{i=1}^{4} i = 1 + 2 + 3 + 4 = 10
result = sum(i for i in range(1, 5))
print(f"Σ_{{i=1}}^{{4}} i = {result}\n")

# Summation of a sequence: Σ_{i=1}^{5} i²
result_sq = sum(i**2 for i in range(1, 6))
print(f"Σ_{{i=1}}^{{5}} i² = 1 + 4 + 9 + 16 + 25 = {result_sq}\n")

# Using numpy for vectorized summation
a = np.array([2, 5, 8, 3, 1])
print(f"a = {a}")
print(f"Σ aᵢ = {np.sum(a)}\n")

# Rule 1: Distributing a constant
c = 3
print("=== Summation Rule: Constant Factor ===")
print(f"c = {c}, a = {a}")
print(f"Σ c·aᵢ = {np.sum(c * a)}")
print(f"c · Σ aᵢ = {c * np.sum(a)}")
print(f"Equal? {np.sum(c * a) == c * np.sum(a)}\n")

# Rule 2: Distributing summation over addition
b = np.array([7, 1, 4, 6, 2])
print("=== Summation Rule: Distribute Over Addition ===")
print(f"a = {a}")
print(f"b = {b}")
print(f"Σ (aᵢ + bᵢ) = {np.sum(a + b)}")
print(f"Σ aᵢ + Σ bᵢ = {np.sum(a) + np.sum(b)}")
print(f"Equal? {np.sum(a + b) == np.sum(a) + np.sum(b)}\n")

# Quantum computing example: completeness relation Σ |i⟩⟨i| = I
ket_0 = np.array([[1], [0]])
ket_1 = np.array([[0], [1]])
completeness = sum(ket @ ket.conj().T for ket in [ket_0, ket_1])
print("=== Completeness Relation ===")
print(f"Σ |i⟩⟨i| = |0⟩⟨0| + |1⟩⟨1| =\n{completeness}  = I ✓")
```
