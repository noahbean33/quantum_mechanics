## Vectors

A long time ago in a country far, far away, there lived an ancient Greek mathematician named Euclid. He wrote a book that defined space using only three dimensions. We will use his vector space to define superposition in quantum computing. Don't be fooled—vector spaces have evolved tremendously since Euclid's days, and our definition of them will evolve too as the book progresses. But for now, we will stick to real numbers, and we'll actually only need two out of the three dimensions Euclid proposed.

To start, we will define a Euclidean vector as being a line segment with a length or magnitude and pointing in a certain direction.

> *Figure 1.1 – Euclidean vector*

Two vectors are equal if they have the same length and direction.

> *Figure 1.2 – Equal vectors*

Vectors can be represented algebraically by their components. The simplest way to do this is to have them start at the origin (the point (0,0)) and use their x and y coordinates.

> *Figure 1.3 – Vectors represented geometrically and algebraically*

You should note that I am using a special notation to label the vectors. It is called bra-ket notation. The appendix has more information on this notation, but for now, we will use a vertical bar or pipe, |, followed by the variable name for the vector and then an angle bracket, ⟩, to denote a vector (for example, |a⟩). The coordinates of our vectors will be enclosed in brackets [ ]. The x coordinate will be on top and the y coordinate on the bottom. Vectors are also called "kets" in this notation—for example, ket a, but for now, we will stick with the name vector.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors |a> and |b>
a = np.array([3, 2])
b = np.array([1, -2])

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Plot individual vectors
ax = axes[0]
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='|a⟩ = [3, 2]')
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='|b⟩ = [1, -2]')
ax.set_xlim(-1, 5)
ax.set_ylim(-3, 4)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Vectors |a⟩ and |b⟩')
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)

# Vector addition geometrically
c = a + b
ax = axes[1]
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='|a⟩')
ax.quiver(a[0], a[1], b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='|b⟩ (shifted)')
ax.quiver(0, 0, c[0], c[1], angles='xy', scale_units='xy', scale=1, color='green', label=f'|c⟩ = |a⟩+|b⟩ = {c}')
ax.set_xlim(-1, 5)
ax.set_ylim(-3, 4)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=8)
ax.set_title('Vector Addition')
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)

# Scalar multiplication
ax = axes[2]
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='|b⟩ = [1, -2]')
b_scaled = 2 * b
ax.quiver(0, 0, b_scaled[0], b_scaled[1], angles='xy', scale_units='xy', scale=1, color='purple', label=f'2|b⟩ = {b_scaled}')
ax.set_xlim(-1, 5)
ax.set_ylim(-5, 2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title('Scalar Multiplication')
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()

print("Vector addition:  |a⟩ + |b⟩ =", a, "+", b, "=", a + b)
print("Scalar multiply:  2·|b⟩ =", 2 * b)
```

### Vector addition

So, it ends up that we can add vectors together both geometrically and algebraically.

> *Figure 1.4 – Vector addition*

As you can see, we can take vectors and move them in the XY-plane as long as we preserve their length and direction. We have taken the vector |b⟩ from our first graph and moved its start position to the end of vector |a⟩. Once we do that, we can draw a third vector |c⟩ that connects the start of |a⟩ and the end of |b⟩ to form their sum. If we look at the coordinates of |c⟩, it is four units in the x direction and zero units in the y direction. This corresponds to the answer we see on the right of Figure 1.4.

We can also do this addition without the help of a graph, as shown on the right of Figure 1.4. Just adding the first components (3 and 1) gives 4, and adding the second components of the vectors (2 and -2) gives 0. Thus, vector addition works both geometrically and algebraically in two dimensions.

## Linear Combinations

Once we have established that we can add our vectors and multiply them by scalars, we can start to talk about linear combinations. Linear combinations are just the scaling and addition of vectors to form new vectors. Let's start with our two vectors we have been working with the whole time, |a⟩ and |b⟩. I want to scale my vector |a⟩ by two to get a new vector |c⟩.

> *Figure 1.6 – |a⟩ scaled by two to produce |c⟩*

Then, I want to take my vector |b⟩ and scale it by two to get a new vector, |d⟩.

> *Figure 1.7 – |b⟩ scaled by two to produce |d⟩*

So, now, we have a vector |c⟩ that is two times |a⟩, and a vector |d⟩ that is two times |b⟩.

Can I add these two new vectors, |c⟩ and |d⟩? Certainly! I will do that, but I will express |e⟩ as a linear combination of |a⟩ and |b⟩.

Vector |e⟩ is a linear combination of vectors |a⟩ and |b⟩!

> *Figure 1.8 – Linear combination*

So, we now have a firm grasp on Euclidean vectors, the algebra you can perform with them, and the concept of a linear combination. We will use that in this next section to describe a quantum phenomenon called superposition.

```python
import numpy as np
import matplotlib.pyplot as plt

# Linear combination: |e⟩ = 2|a⟩ + 2|b⟩
a = np.array([3, 2])
b = np.array([1, -2])
c = 2 * a
d = 2 * b
e = c + d  # = 2*a + 2*b

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Step 1: Scale |a⟩ by 2
ax = axes[0]
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='|a⟩', linewidth=1.5)
ax.quiver(0, 0, c[0], c[1], angles='xy', scale_units='xy', scale=1, color='cyan', label='|c⟩ = 2|a⟩', linewidth=2)
ax.set_xlim(-1, 8); ax.set_ylim(-5, 6)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend(); ax.set_title('Step 1: Scale |a⟩ by 2')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# Step 2: Scale |b⟩ by 2
ax = axes[1]
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='|b⟩', linewidth=1.5)
ax.quiver(0, 0, d[0], d[1], angles='xy', scale_units='xy', scale=1, color='orange', label='|d⟩ = 2|b⟩', linewidth=2)
ax.set_xlim(-1, 8); ax.set_ylim(-5, 6)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend(); ax.set_title('Step 2: Scale |b⟩ by 2')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# Step 3: Add them → linear combination
ax = axes[2]
ax.quiver(0, 0, c[0], c[1], angles='xy', scale_units='xy', scale=1, color='cyan', label='|c⟩ = 2|a⟩', alpha=0.5)
ax.quiver(c[0], c[1], d[0], d[1], angles='xy', scale_units='xy', scale=1, color='orange', label='|d⟩ = 2|b⟩', alpha=0.5)
ax.quiver(0, 0, e[0], e[1], angles='xy', scale_units='xy', scale=1, color='green', label=f'|e⟩ = 2|a⟩+2|b⟩ = {e}', linewidth=2.5)
ax.set_xlim(-1, 9); ax.set_ylim(-5, 6)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend(fontsize=8); ax.set_title('Step 3: Linear Combination')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()

print(f"|e⟩ = 2·|a⟩ + 2·|b⟩ = 2·{a} + 2·{b} = {e}")
```

## Superposition

Superposition can be a very imposing term, so before we delve into it, let's take a step back and talk about the computers we use today. In quantum computing, we call these computers "classical computers" to distinguish them from quantum computers. Classical computers use binary digits—or bits, for short—to store ones and zeros. These ones and zeros can represent anything, from truth values to characters to pixel values on a screen! They are physically implemented using any two-state device such as an electrical switch that is either on or off.

A quantum bit, or qubit for short, is the analogous building block of quantum computers. They are implemented by anything that demonstrates quantum phenomena, which means they are very, very small. In the following screenshot, we show how a property of an electron—namely spin—can be used to represent a one or zero of a qubit:

> *Figure 1.9 – Pair of electrons with a spin labeled 1 and 0*

Physicists use mathematics to model quantum phenomena, and guess what they use to model the state of a quantum particle? That's right! Vectors! Quantum computer scientists have taken two of these states and labeled them as the canonical one and zero for qubits.

> *Figure 1.10 – Zero and one states*

As you can see, the zero and one states are just vectors on the x and y axes with a length of one unit each. When you combine a lot of ones and zeros in classical computing, wonderful, complex things can be done. The same is true of the zero and one state of qubits in quantum computing.

> **Greek Letters:** Mathematicians and physicists love Greek letters, and they have found their way into quantum computing in several places. The Greek letter "Psi", ψ, is often used to represent the state of a qubit. The Greek letters "alpha", α, and "beta", β, are used to represent numbers or scalars.

While qubits can represent a one or a zero, they have a superpower in that they can represent a combination of a zero and one as well! "How?" you might ask. Well, this is where superposition comes in. Understanding it is actually quite simple from a mathematical standpoint. In fact, you already know what it is! It's just a fancy way of saying that a qubit is in a linear combination of states.

> *Figure 1.11 – Definition of |e⟩*

If we replace those letters and numbers with the Greek letters and the zero and one states we just introduced, we get an equation like this:

> *Figure 1.12 – Greek letters being transposed onto a linear combination equation*

The bottom equation represents a qubit in the state |ψ⟩, which is a superposition of the states zero and one! You now know what superposition is mathematically! This, by the way, is the only way that counts because math is the language of physics and, therefore, quantum computing.

```python
import numpy as np
import matplotlib.pyplot as plt

# Qubit basis states
ket_0 = np.array([1, 0])
ket_1 = np.array([0, 1])

# Superposition state: |ψ⟩ = α|0⟩ + β|1⟩ with α = β = 1/√2
alpha = 1 / np.sqrt(2)
beta = 1 / np.sqrt(2)
psi = alpha * ket_0 + beta * ket_1

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot the basis states and superposition vector
ax = axes[0]
ax.quiver(0, 0, ket_0[0], ket_0[1], angles='xy', scale_units='xy', scale=1, color='blue', linewidth=2, label='|0⟩')
ax.quiver(0, 0, ket_1[0], ket_1[1], angles='xy', scale_units='xy', scale=1, color='red', linewidth=2, label='|1⟩')
ax.quiver(0, 0, psi[0], psi[1], angles='xy', scale_units='xy', scale=1, color='purple', linewidth=2.5, label=f'|ψ⟩ = {alpha:.3f}|0⟩ + {beta:.3f}|1⟩')
ax.set_xlim(-0.2, 1.3); ax.set_ylim(-0.2, 1.3)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend(fontsize=10)
ax.set_title('Qubit Superposition as a Vector')
ax.set_xlabel('|0⟩ component'); ax.set_ylabel('|1⟩ component')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# Measurement probabilities
prob_0 = np.abs(alpha)**2
prob_1 = np.abs(beta)**2
ax = axes[1]
bars = ax.bar(['|0⟩', '|1⟩'], [prob_0, prob_1], color=['blue', 'red'], alpha=0.7, edgecolor='black')
ax.set_ylim(0, 1.1)
ax.set_ylabel('Probability')
ax.set_title('Measurement Probabilities')
for bar, prob in zip(bars, [prob_0, prob_1]):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02, f'{prob:.1%}', ha='center', fontsize=12)

plt.tight_layout()
plt.show()

print(f"α = 1/√2 ≈ {alpha:.4f}")
print(f"β = 1/√2 ≈ {beta:.4f}")
print(f"|α|² = {prob_0:.4f}  → probability of measuring |0⟩")
print(f"|β|² = {prob_1:.4f}  → probability of measuring |1⟩")
print(f"|α|² + |β|² = {prob_0 + prob_1:.4f}  (must equal 1)")

# Simulate measurements
np.random.seed(42)
n_measurements = 10000
outcomes = np.random.choice([0, 1], size=n_measurements, p=[prob_0, prob_1])
print(f"\nSimulating {n_measurements} measurements:")
print(f"  Got |0⟩: {np.sum(outcomes == 0)} times ({np.mean(outcomes == 0):.1%})")
print(f"  Got |1⟩: {np.sum(outcomes == 1)} times ({np.mean(outcomes == 1):.1%})")
```
