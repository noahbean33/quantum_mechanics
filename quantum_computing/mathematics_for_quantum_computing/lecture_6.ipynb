## Three Forms, One Number

There are three main ways of representing a complex number:

1. Cartesian form (aka the general form)
2. Polar form
3. Exponential form

### Definition of complex numbers

A complex number is a number that can be expressed as:

$$z = a + bi \quad \text{(1)}$$

where a and b are real numbers and i is the imaginary unit, defined as:

$$i = \sqrt{-1}, \quad i^2 = -1$$

The real part is denoted by Re(z) and the imaginary part by Im(z). For z defined in Equation (1), Re(z) = a and Im(z) = b. Two complex numbers are equal if, and only if, their real and imaginary parts are equal.

It is important to remember that ℝ is a subset of ℂ. Hence, if Im(z) = 0, then z is also a real number.

## Cartesian Form

We used the Cartesian form to define a complex number. We can represent complex numbers on a Cartesian coordinate system since a and b are just real numbers. We replace the x axis with Re(z) and the y axis with Im(z):

> *Figure 6.1 – The complex plane*

> *Figure 6.2 – Complex numbers on the complex plane*

### Addition

Addition is rather easy for complex numbers; just add Re(z) and Im(z) of the two numbers together. For two complex numbers:

$$z_1 = a + bi, \quad z_2 = c + di$$

$$z_1 + z_2 = (a + c) + (b + d)i$$
$$z_1 - z_2 = (a - c) + (b - d)i$$

> *Figure 6.3 – Vector addition and scalar multiplication*

### Multiplication

Using the FOIL method:

> *Figure 6.4 – FOIL method illustrated*

The definition of multiplication:

$$z_1 \cdot z_2 = (a + bi)(c + di) = (ac - bd) + (ad + bc)i$$

**Exercise 1:** Compute multiplication problems.

### Complex conjugate

The complex conjugate of a complex number a + bi is a – bi. It is written as z* for a complex number z:

$$z = a + bi \implies z^* = a - bi$$

Complex conjugation on the complex plane is just a reflection of the real axis.

> *Figure 6.5 – Complex vector reflected on the real axis*

### Absolute value or modulus

The absolute value or modulus of a complex number z, denoted |z|, is:

$$|z| = \sqrt{z \cdot z^*} = \sqrt{a^2 + b^2}$$

**Exercise 2:** Compute absolute values.

### Division

Division involves "rationalizing the denominator." Multiply numerator and denominator by the complex conjugate of the denominator:

$$\frac{z_1}{z_2} = \frac{(a + bi)}{(c + di)} \cdot \frac{(c - di)}{(c - di)} = \frac{(ac + bd) + (bc - ad)i}{c^2 + d^2}$$

### Powers of i

The positive powers of i follow this pattern:

$$i^1 = i, \quad i^2 = -1, \quad i^3 = -i, \quad i^4 = 1, \quad i^5 = i, \ldots$$

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Complex numbers on the complex plane ---
z1 = 3 + 2j
z2 = -1 + 4j
z3 = 2 - 3j

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot complex numbers
ax = axes[0]
for z, label, color in [(z1, 'z₁=3+2i', 'blue'), (z2, 'z₂=-1+4i', 'red'), (z3, 'z₃=2-3i', 'green')]:
    ax.quiver(0, 0, z.real, z.imag, angles='xy', scale_units='xy', scale=1, color=color, linewidth=2, label=label)
    ax.plot(z.real, z.imag, 'o', color=color, markersize=6)
ax.set_xlim(-3, 5); ax.set_ylim(-4, 5)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend()
ax.set_xlabel('Re(z)'); ax.set_ylabel('Im(z)')
ax.set_title('Complex Numbers on the Complex Plane')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# Complex conjugate
ax = axes[1]
z = 2 + 3j
z_conj = np.conj(z)
ax.quiver(0, 0, z.real, z.imag, angles='xy', scale_units='xy', scale=1, color='blue', linewidth=2, label=f'z = {z}')
ax.quiver(0, 0, z_conj.real, z_conj.imag, angles='xy', scale_units='xy', scale=1, color='red', linewidth=2, label=f'z* = {z_conj}')
ax.axhline(y=0, color='k', linewidth=1.5, linestyle='--', alpha=0.5)
ax.set_xlim(-1, 4); ax.set_ylim(-4, 4)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend()
ax.set_xlabel('Re(z)'); ax.set_ylabel('Im(z)')
ax.set_title('Complex Conjugate (reflection over real axis)')

# Powers of i
ax = axes[2]
powers = [1j**n for n in range(8)]
for n, p in enumerate(powers):
    ax.plot(p.real, p.imag, 'o', markersize=10, color=plt.cm.viridis(n/8))
    ax.annotate(f'i^{n}', (p.real, p.imag), textcoords="offset points", xytext=(8, 5), fontsize=9)
circle = plt.Circle((0, 0), 1, fill=False, color='gray', linestyle='--', alpha=0.5)
ax.add_patch(circle)
ax.set_xlim(-1.5, 1.5); ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
ax.set_xlabel('Re'); ax.set_ylabel('Im')
ax.set_title('Powers of i (cycle of 4)')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()

# Arithmetic demo
print("=== Complex Arithmetic ===")
print(f"z₁ = {z1},  z₂ = {z2}")
print(f"z₁ + z₂ = {z1 + z2}")
print(f"z₁ × z₂ = {z1 * z2}")
print(f"|z₁| = {abs(z1):.4f}")
print(f"z₁* = {np.conj(z1)}")
print(f"z₁ / z₂ = {z1/z2:.4f}")
```

## Polar Form

The polar form is based on polar coordinates. The rest of the chapter is heavy on trigonometry and uses radians for all angles.

### Polar coordinates

Polar coordinates are another way of representing points in ℝ². We represent a point with two coordinates called r and θ.

> *Figure 6.6 – Polar coordinates on a graph*

$$r = \sqrt{x^2 + y^2} \quad \text{(2)}$$

$$\theta = \arctan\left(\frac{y}{x}\right) \quad \text{(3)}$$

**Exercise 3:** Convert points to polar coordinates.

### Defining complex numbers in polar form

> *Figure 6.8 – Polar coordinates of the complex number z*

The r in the hypotenuse represents the modulus of our complex number:

$$r = |z| = \sqrt{a^2 + b^2}$$

The angle θ is called the **argument** of the complex number (or the **phase** in quantum computing).

$$\text{Re}(z) = r\cos\theta, \quad \text{Im}(z) = r\sin\theta$$

In polar form:

$$z = r(\cos\theta + i\sin\theta)$$

### Multiplication and division in polar form

Given two complex numbers in polar form:

$$z_1 \cdot z_2 = r_1 r_2 [\cos(\theta_1 + \theta_2) + i\sin(\theta_1 + \theta_2)] \quad \text{(4)}$$

Just add the angles and multiply the moduli for multiplication. Subtract the angles and divide the moduli for division.

### De Moivre's theorem

$$z^n = r^n(\cos(n\theta) + i\sin(n\theta))$$

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Polar and Exponential forms ---
# Convert Cartesian to Polar
z = 1 + 1j
r = abs(z)
theta = np.angle(z)

print(f"z = {z}")
print(f"Polar:       r = {r:.4f}, θ = {theta:.4f} rad = {np.degrees(theta):.1f}°")
print(f"Polar form:  z = {r:.4f}(cos({theta:.4f}) + i·sin({theta:.4f}))")
print(f"Exponential: z = {r:.4f} · e^(i·{theta:.4f})")
print(f"Verify:      {r * np.exp(1j * theta):.4f}\n")

# Euler's formula visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# e^(iθ) traces a unit circle
ax = axes[0]
thetas = np.linspace(0, 2*np.pi, 200)
points = np.exp(1j * thetas)
ax.plot(points.real, points.imag, 'b-', linewidth=2)
# Mark special angles
special = {'0': 0, 'π/4': np.pi/4, 'π/2': np.pi/2, 'π': np.pi, '3π/2': 3*np.pi/2}
for label, th in special.items():
    p = np.exp(1j * th)
    ax.plot(p.real, p.imag, 'ro', markersize=8)
    ax.annotate(f'θ={label}', (p.real, p.imag), textcoords="offset points", xytext=(10, 5), fontsize=9)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
ax.set_xlabel('Re'); ax.set_ylabel('Im')
ax.set_title('Euler\'s Formula: e^(iθ) = cos(θ) + i·sin(θ)\nTraces the unit circle')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# Multiplication in polar form = add angles, multiply moduli
ax = axes[1]
z1 = 2 * np.exp(1j * np.pi/6)    # r=2, θ=30°
z2 = 1.5 * np.exp(1j * np.pi/4)  # r=1.5, θ=45°
z_prod = z1 * z2                   # r=3, θ=75°
for z, label, color in [(z1, f'z₁: r={abs(z1):.1f}, θ=30°', 'blue'),
                          (z2, f'z₂: r={abs(z2):.1f}, θ=45°', 'red'),
                          (z_prod, f'z₁z₂: r={abs(z_prod):.1f}, θ=75°', 'purple')]:
    ax.quiver(0, 0, z.real, z.imag, angles='xy', scale_units='xy', scale=1, color=color, linewidth=2, label=label)
ax.set_xlim(-1, 4); ax.set_ylim(-1, 4)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend(fontsize=9)
ax.set_title('Multiplication: multiply moduli, add angles')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()

# Euler's identity
result = np.exp(1j * np.pi) + 1
print(f"Euler's Identity: e^(iπ) + 1 = {result:.2e}  ≈ 0  ✓")
```

## The Most Beautiful Equation in Mathematics

In 1748, Leonhard Euler (pronounced "oy-lr") published his most famous formula, **Euler's Formula**:

$$e^{i\theta} = \cos\theta + i\sin\theta$$

Substituting θ = π gives the most beautiful equation in math, **Euler's identity**:

$$e^{i\pi} + 1 = 0$$

It combines the operations of addition, multiplication, and exponentiation with 0 (the additive identity), 1 (the multiplicative identity), i (the imaginary unit), and two of the most important mathematical constants, e and π. This equation is also integral to quantum computing. You will see $e^{i\theta}$ all over the place in quantum computing.

## Exponential Form

Complex numbers written in terms of $e^{i\theta}$ are said to be in the exponential form:

$$z = re^{i\theta}$$

The exponential form is very close to polar form, but now you have θ in one place instead of two!

**Exercise 4:** Express complex numbers in exponential form.

### Conjugation

For complex numbers in exponential form, just change the sign of the angle:

$$z = re^{i\theta} \implies z^* = re^{-i\theta}$$

### Multiplication

Multiplication and division are even easier in exponential form:

$$z_1 \cdot z_2 = r_1 r_2 e^{i(\theta_1 + \theta_2)}$$

$$\frac{z_1}{z_2} = \frac{r_1}{r_2} e^{i(\theta_1 - \theta_2)}$$

## Conjugate Transpose of a Matrix

Since we now have the definition of the complex conjugate of a number, let's go over the conjugate transpose of a matrix.

The transpose converts rows into columns and columns into rows. The conjugate of a matrix is the conjugation of every entry.

The **conjugate transpose** of a matrix A is:

$$A^\dagger = (A^*)^T$$

The cross symbol (†) is pronounced "dagger," so "A dagger" means the conjugate transpose of A.

The conjugate transpose also goes under the names **Hermitian conjugate** and **adjoint matrix**.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- Conjugate Transpose (Dagger) ---
A = np.array([[1+2j, 3-1j], [4j, 2+5j]])
A_dagger = A.conj().T

print("=== Conjugate Transpose ===")
print(f"A =\n{A}\n")
print(f"A* (conjugate) =\n{A.conj()}\n")
print(f"A^T (transpose) =\n{A.T}\n")
print(f"A† = (A*)^T =\n{A_dagger}\n")

# --- Bloch Sphere ---
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Draw the sphere wireframe
u_mesh = np.linspace(0, 2 * np.pi, 30)
v_mesh = np.linspace(0, np.pi, 20)
x_s = np.outer(np.cos(u_mesh), np.sin(v_mesh))
y_s = np.outer(np.sin(u_mesh), np.sin(v_mesh))
z_s = np.outer(np.ones(u_mesh.size), np.cos(v_mesh))
ax.plot_wireframe(x_s, y_s, z_s, color='lightgray', alpha=0.3, linewidth=0.5)

# Draw axes
ax.quiver(0, 0, 0, 0, 0, 1.3, color='k', arrow_length_ratio=0.05, linewidth=1)
ax.quiver(0, 0, 0, 1.3, 0, 0, color='k', arrow_length_ratio=0.05, linewidth=1)
ax.quiver(0, 0, 0, 0, 1.3, 0, color='k', arrow_length_ratio=0.05, linewidth=1)
ax.text(0, 0, 1.4, '|0⟩', fontsize=12, fontweight='bold', color='blue')
ax.text(0, 0, -1.4, '|1⟩', fontsize=12, fontweight='bold', color='red')
ax.text(1.4, 0, 0, 'x', fontsize=10)
ax.text(0, 1.4, 0, 'y', fontsize=10)

# Plot specific qubit states
states = {
    '|0⟩': (0, 0),                    # North pole
    '|1⟩': (np.pi, 0),                # South pole
    '|+⟩': (np.pi/2, 0),              # +x
    '|-⟩': (np.pi/2, np.pi),          # -x
    '|+i⟩': (np.pi/2, np.pi/2),       # +y
    '|-i⟩': (np.pi/2, 3*np.pi/2),     # -y
}
colors_map = {'|0⟩': 'blue', '|1⟩': 'red', '|+⟩': 'green', '|-⟩': 'orange', '|+i⟩': 'purple', '|-i⟩': 'brown'}

for name, (theta, phi) in states.items():
    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.sin(phi)
    z = np.cos(theta)
    ax.scatter(x, y, z, s=80, color=colors_map[name], zorder=5)
    ax.text(x*1.15, y*1.15, z*1.15, name, fontsize=10, color=colors_map[name], fontweight='bold')

# Plot a general state
theta_gen, phi_gen = np.pi/3, np.pi/4
x_g = np.sin(theta_gen) * np.cos(phi_gen)
y_g = np.sin(theta_gen) * np.sin(phi_gen)
z_g = np.cos(theta_gen)
ax.quiver(0, 0, 0, x_g, y_g, z_g, color='magenta', linewidth=2.5, arrow_length_ratio=0.08)
ax.text(x_g*1.1, y_g*1.1, z_g*1.1, f'|ψ⟩\nθ={np.degrees(theta_gen):.0f}°\nφ={np.degrees(phi_gen):.0f}°', 
        fontsize=9, color='magenta')

ax.set_xlim(-1.5, 1.5); ax.set_ylim(-1.5, 1.5); ax.set_zlim(-1.5, 1.5)
ax.set_title('Bloch Sphere', fontsize=14, fontweight='bold')
ax.set_box_aspect([1,1,1])
plt.tight_layout()
plt.show()

# Qubit state from Bloch angles
alpha_bloch = np.cos(theta_gen / 2)
beta_bloch = np.exp(1j * phi_gen) * np.sin(theta_gen / 2)
print(f"\nBloch sphere state: θ={np.degrees(theta_gen):.0f}°, φ={np.degrees(phi_gen):.0f}°")
print(f"|ψ⟩ = {alpha_bloch:.4f}|0⟩ + ({beta_bloch:.4f})|1⟩")
print(f"|α|² + |β|² = {abs(alpha_bloch)**2 + abs(beta_bloch)**2:.4f}  (= 1 ✓)")
```

## Bloch Sphere

The Bloch sphere, named after Felix Bloch, is a way to visualize a single qubit. A qubit can be represented as:

$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$$

where α and β are complex numbers. Since complex numbers take two real numbers to represent, it looks like we need four real numbers. Let's reduce this.

Replacing α and β with exponential form and factoring out a global phase (which has no measurable effect):

$$|\psi\rangle = r_\alpha|0\rangle + r_\beta e^{i\varphi}|1\rangle$$

Using the constraint that all probabilities must add up to one ($|\alpha|^2 + |\beta|^2 = 1$), we can parameterize with a single angle θ:

$$r_\alpha = \cos\frac{\theta}{2}, \quad r_\beta = \sin\frac{\theta}{2}$$

The final result:

$$|\psi\rangle = \cos\frac{\theta}{2}|0\rangle + e^{i\varphi}\sin\frac{\theta}{2}|1\rangle$$

We're down to only **two dimensions** (θ and φ). Using these two angles, we can represent the qubit state on a unit sphere:

> *Figure 6.10 – Bloch sphere*

θ is like the latitude on a map: zero on the north pole (|0⟩ state), π/2 on the equator, and π on the south pole (|1⟩ state).
