## The Inner Product

An inner product can actually be any function that follows a few properties, but we are going to zero in on one definition of the inner product that we will use in quantum computing.

Dirac defined it with a bra and ket, calling it a **bracket**:

$$\langle x | y \rangle$$

If we define a bra to be the conjugate transpose of its corresponding ket, so that if:

$$|x\rangle = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}$$

Then:

$$\langle x| = \begin{bmatrix} x_1^* & x_2^* & \cdots & x_n^* \end{bmatrix}$$

We can then define a bracket as just matrix multiplication! The bra ⟨x| is a linear functional — it will take any vector |y⟩ and give you a scalar according to the inner product formula.

An important property:

$$\langle y|x\rangle = \langle x|y\rangle^*$$

Since bras are linear functionals, brackets are left-distributive (α and β are scalars):

$$\langle x|(\alpha|y\rangle + \beta|z\rangle) = \alpha\langle x|y\rangle + \beta\langle x|z\rangle \quad \text{(1)}$$

And they are also right-distributive.

```python
import numpy as np

# --- Inner Product (Bracket) ---
# |x⟩ and |y⟩ as complex column vectors
x = np.array([[1+1j], [2-1j]])
y = np.array([[3], [1j]])

# ⟨x| = conjugate transpose of |x⟩
bra_x = x.conj().T
inner = bra_x @ y

print("=== Inner Product ===")
print(f"|x⟩ =\n{x}\n")
print(f"|y⟩ =\n{y}\n")
print(f"⟨x| = (|x⟩)† = {bra_x}")
print(f"⟨x|y⟩ = {inner[0,0]}\n")

# Property: ⟨y|x⟩ = ⟨x|y⟩*
bra_y = y.conj().T
inner_yx = bra_y @ x
print(f"⟨y|x⟩ = {inner_yx[0,0]}")
print(f"⟨x|y⟩* = {np.conj(inner[0,0])}")
print(f"⟨y|x⟩ == ⟨x|y⟩*? {np.isclose(inner_yx[0,0], np.conj(inner[0,0]))}\n")

# Norm: ||x|| = sqrt(⟨x|x⟩)
norm_x = np.sqrt((bra_x @ x)[0,0]).real
print(f"||x|| = √⟨x|x⟩ = {norm_x:.4f}")
print(f"np.linalg.norm check: {np.linalg.norm(x):.4f}")
```

## Orthonormality

### The norm

We can define a metric on our vector spaces called the **norm**, denoted ∥x∥. In two- and three-dimensional Euclidean spaces, it is often called the length of a vector.

$$\|x\| = \sqrt{\langle x|x\rangle}$$

### Normalization and unit vectors

A **unit vector** has norm equal to one. We divide a vector by its norm in a process called **normalization**:

$$|u\rangle = \frac{|v\rangle}{\|v\|}$$

### Orthogonality

Orthogonality generalizes the concept of perpendicular vectors to higher dimensions of ℂⁿ. If two vectors are orthogonal, their inner product is 0:

$$\langle x|y\rangle = 0 \implies |x\rangle \perp |y\rangle$$

### Orthonormal vectors

If we combine orthogonality and normalization, we get **orthonormal vectors**.

The computational basis vectors |0⟩ and |1⟩ are orthonormal:
- Their inner product is zero (orthogonal)
- Their norms are one (normalized)

Another important set of orthonormal vectors in quantum computing:

$$|+\rangle = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix}, \quad |-\rangle = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \end{bmatrix}$$

### The Kronecker delta function

$$\delta_{ij} = \begin{cases} 1 & \text{if } i = j \\ 0 & \text{if } i \neq j \end{cases}$$

For orthonormal basis vectors:

$$\langle b_i | b_j \rangle = \delta_{ij}$$

```python
import numpy as np

# --- Orthonormality ---
ket_0 = np.array([[1], [0]])
ket_1 = np.array([[0], [1]])
ket_plus = (1/np.sqrt(2)) * np.array([[1], [1]])
ket_minus = (1/np.sqrt(2)) * np.array([[1], [-1]])

def inner(a, b):
    return (a.conj().T @ b)[0, 0]

def norm(a):
    return np.sqrt(inner(a, a).real)

print("=== Orthonormality of Computational Basis ===")
print(f"⟨0|0⟩ = {inner(ket_0, ket_0):.1f}  (normalized)")
print(f"⟨1|1⟩ = {inner(ket_1, ket_1):.1f}  (normalized)")
print(f"⟨0|1⟩ = {inner(ket_0, ket_1):.1f}  (orthogonal)")
print(f"⟨1|0⟩ = {inner(ket_1, ket_0):.1f}  (orthogonal)")

print(f"\n=== |+⟩ and |−⟩ Basis ===")
print(f"⟨+|+⟩ = {inner(ket_plus, ket_plus):.1f}  (normalized)")
print(f"⟨-|-⟩ = {inner(ket_minus, ket_minus):.1f}  (normalized)")
print(f"⟨+|-⟩ = {inner(ket_plus, ket_minus):.1f}  (orthogonal)")

# Normalization example
v = np.array([[3], [4]])
v_normalized = v / norm(v)
print(f"\n=== Normalization ===")
print(f"|v⟩ = {v.flatten()},  ||v|| = {norm(v):.4f}")
print(f"|v̂⟩ = |v⟩/||v|| = {np.round(v_normalized.flatten(), 4)}")
print(f"||v̂|| = {norm(v_normalized):.4f}  (unit vector ✓)")

# Outer product
print(f"\n=== Outer Product ===")
outer_00 = ket_0 @ ket_0.conj().T
outer_11 = ket_1 @ ket_1.conj().T
print(f"|0⟩⟨0| =\n{outer_00}\n")
print(f"|1⟩⟨1| =\n{outer_11}\n")
print(f"Completeness: |0⟩⟨0| + |1⟩⟨1| =\n{outer_00 + outer_11}  = I ✓")
```

## The Outer Product

The outer product is matrix multiplication of two vectors that produces a **matrix** (unlike the inner product which produces a scalar):

$$|u\rangle\langle v|$$

We have an m × 1 matrix and a 1 × p matrix, producing an m × p matrix.

The outer product is **not commutative**: $|u\rangle\langle v| \neq |v\rangle\langle u|$

But the results are transposes of each other:

$$(|u\rangle\langle v|)^T = |v\rangle\langle u|$$

The outer product is distributive over addition:

$$(|u\rangle + |v\rangle)\langle w| = |u\rangle\langle w| + |v\rangle\langle w|$$

$$|u\rangle(\langle v| + \langle w|) = |u\rangle\langle v| + |u\rangle\langle w| \quad \text{(2)}$$

## Operators

Linear operators are linear transformations that map vectors from and to the same vector space. They are represented by square matrices.

The X̂ operator transforms the zero and one states:

$$\hat{X}|0\rangle = |1\rangle, \quad \hat{X}|1\rangle = |0\rangle$$

In the computational basis {|0⟩, |1⟩}:

$$X = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$$

In the {|+⟩, |-⟩} basis:

$$X_{+,-} = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$$

### Representing an operator using the outer product

Since the outer product results in a matrix, we can represent operators as outer products. For A = |u⟩⟨v|:

$$A|x\rangle = |u\rangle\langle v|x\rangle = \langle v|x\rangle |u\rangle \quad \text{(4)}$$

The resulting ket will be proportional to the ket used in the outer product.

### The completeness relation

Starting with an orthonormal basis B: $B = \{|b_1\rangle, |b_2\rangle, \ldots, |b_n\rangle\}$

The identity operator can be written as:

$$\hat{I} = \sum_i |b_i\rangle\langle b_i| \quad \text{(5)}$$

This is used over and over again in quantum computing to manipulate bra-ket expressions.

### The adjoint of an operator

For every ket |x⟩, there is a bra ⟨x|. If an operator Â acts on |x⟩ to give |y⟩:

$$\hat{A}|x\rangle = |y\rangle$$

The adjoint of Â acts on bras:

$$\langle x|\hat{A}^\dagger = \langle y|$$

Rules for the adjoint:
- The adjoint of a scalar is its complex conjugate: $(\alpha)^\dagger = \alpha^*$
- Distributes across scalar multiplication: $(\alpha\hat{A})^\dagger = \alpha^*\hat{A}^\dagger$
- Bras and kets are adjoints of each other: $(|x\rangle)^\dagger = \langle x|$
- Double adjoint returns the original: $(\hat{A}^\dagger)^\dagger = \hat{A}$
- Anti-commutes across multiplication: $(\hat{A}\hat{B})^\dagger = \hat{B}^\dagger\hat{A}^\dagger$
- Distributes across sums: $(\hat{A} + \hat{B})^\dagger = \hat{A}^\dagger + \hat{B}^\dagger$

For an outer product:

$$(|u\rangle\langle v|)^\dagger = |v\rangle\langle u|$$

When represented as a matrix on an orthonormal basis, the adjoint is the conjugate transpose:

$$A^\dagger = (A^*)^T$$

```python
import numpy as np

# --- Types of Operators ---
I2 = np.eye(2)

# Pauli matrices
X = np.array([[0, 1], [1, 0]])
Y = np.array([[0, -1j], [1j, 0]])
Z = np.array([[1, 0], [0, -1]])
H = (1/np.sqrt(2)) * np.array([[1, 1], [1, -1]])

def is_hermitian(M):
    return np.allclose(M, M.conj().T)

def is_unitary(M):
    return np.allclose(M.conj().T @ M, np.eye(M.shape[0]))

def is_normal(M):
    return np.allclose(M @ M.conj().T, M.conj().T @ M)

def is_projection(M):
    return is_hermitian(M) and np.allclose(M @ M, M)

print("=== Operator Type Checks ===\n")
for name, M in [('X (Pauli)', X), ('Y (Pauli)', Y), ('Z (Pauli)', Z), ('H (Hadamard)', H)]:
    evals = np.linalg.eigvals(M)
    print(f"{name}:")
    print(f"  Hermitian? {is_hermitian(M)}   Unitary? {is_unitary(M)}   Normal? {is_normal(M)}")
    print(f"  Eigenvalues: {np.round(evals, 4)}")
    print(f"  |λ| = {np.round(np.abs(evals), 4)}  (all = 1 for unitary)")
    print()

# Projection operator
P0 = np.array([[1, 0], [0, 0]])  # |0⟩⟨0|
P1 = np.array([[0, 0], [0, 1]])  # |1⟩⟨1|
print("=== Projection Operators ===")
print(f"|0⟩⟨0| =\n{P0}")
print(f"  Hermitian? {is_hermitian(P0)}   P² = P? {np.allclose(P0 @ P0, P0)}")
print(f"  Eigenvalues: {np.linalg.eigvals(P0)}  (only 0 and 1)\n")

# Unitary preserves inner product
ket_0 = np.array([[1], [0]])
ket_1 = np.array([[0], [1]])
print("=== Unitary Preserves Inner Product ===")
ip_before = (ket_0.conj().T @ ket_1)[0,0]
ip_after = ((H @ ket_0).conj().T @ (H @ ket_1))[0,0]
print(f"⟨0|1⟩ = {ip_before}")
print(f"⟨0|H†H|1⟩ = {np.round(ip_after, 10)}")
print(f"Preserved? {np.isclose(ip_before, ip_after)}")
```

## Types of Operators

### Normal operators

Normal operators commute with their adjoint:

$$\hat{A}\hat{A}^\dagger = \hat{A}^\dagger\hat{A} \quad \text{(6)}$$

Properties of normal operators/matrices:
- They are diagonalizable
- Their eigenvalues are the conjugates of the eigenvalues of their adjoints
- Eigenvectors associated with different eigenvalues are orthogonal
- A vector space can be defined by an orthogonal basis set of these eigenvectors

### Hermitian operators

A Hermitian (self-adjoint) operator is equal to its adjoint:

$$\hat{A} = \hat{A}^\dagger$$

Additional property: **All their eigenvalues are real numbers.** Hermitian operators play an important part in quantum computing — all measurements are done via a Hermitian operator.

### Unitary operators

Unitary operators describe the evolution of a quantum state — **all gates in quantum computing are unitary**:

$$\hat{U}^\dagger\hat{U} = \hat{U}\hat{U}^\dagger = \hat{I}$$

Properties:
- Eigenvalues are complex numbers of modulus one: $|\lambda| = 1$
- They preserve the inner product: $\langle u|\hat{U}^\dagger\hat{U}|v\rangle = \langle u|v\rangle$
- They preserve the norm of transformed vectors

For unitary matrices:

$$U^\dagger U = UU^\dagger = I$$

### Projection operators

In quantum computing, projection operators are defined as:

$$\hat{P} = \hat{P}^\dagger = \hat{P}^2 \quad \text{(7, 8)}$$

Projection operators are Hermitian. If two projection operators commute, their product is also a projection operator.

To project onto an m-dimensional subspace with orthonormal basis {|0⟩, |1⟩, …, |m⟩}:

$$\hat{P} = \sum_{i=0}^{m} |i\rangle\langle i|$$

The only eigenvalues a projection operator can have are **zero and one**.

### Positive operators

Positive operators are Hermitian. An operator Â is **positive definite** if:

$$\langle x|\hat{A}|x\rangle > 0 \quad \text{(9)}$$

An operator Â is **positive semidefinite** if:

$$\langle x|\hat{A}|x\rangle \geq 0$$

All eigenvalues of positive operators are non-negative.

```python
import numpy as np

# --- Tensor Products (Kronecker Product) ---
ket_0 = np.array([[1], [0]])
ket_1 = np.array([[0], [1]])

# Two-qubit basis states using Kronecker product
ket_00 = np.kron(ket_0, ket_0)
ket_01 = np.kron(ket_0, ket_1)
ket_10 = np.kron(ket_1, ket_0)
ket_11 = np.kron(ket_1, ket_1)

print("=== Two-Qubit Basis States (Tensor Product) ===")
for name, state in [('|00⟩', ket_00), ('|01⟩', ket_01), ('|10⟩', ket_10), ('|11⟩', ket_11)]:
    print(f"{name} = {state.flatten()}")

# Kronecker product of matrices (tensor product of operators)
X = np.array([[0, 1], [1, 0]])
I2 = np.eye(2)
XI = np.kron(X, I2)  # X on first qubit, I on second
IX = np.kron(I2, X)  # I on first qubit, X on second

print(f"\n=== Tensor Product of Operators ===")
print(f"X ⊗ I =\n{XI.astype(int)}\n")
print(f"I ⊗ X =\n{IX.astype(int)}\n")

# Apply X⊗I to |00⟩ → should flip first qubit → |10⟩
result = XI @ ket_00
print(f"(X⊗I)|00⟩ = {result.flatten().astype(int)}  = |10⟩ ✓")

# Apply I⊗X to |00⟩ → should flip second qubit → |01⟩
result2 = IX @ ket_00
print(f"(I⊗X)|00⟩ = {result2.flatten().astype(int)}  = |01⟩ ✓")

# Inner product of composite vectors
print(f"\n=== Inner Product in Tensor Product Space ===")
ip = (ket_01.conj().T @ ket_10)[0,0]
print(f"⟨01|10⟩ = {ip:.0f}  (orthogonal)")
ip2 = (ket_01.conj().T @ ket_01)[0,0]
print(f"⟨01|01⟩ = {ip2:.0f}  (normalized)")

# Verify: ⟨01|10⟩ = ⟨0|1⟩ · ⟨1|0⟩
ip_factor = (ket_0.conj().T @ ket_1)[0,0] * (ket_1.conj().T @ ket_0)[0,0]
print(f"⟨0|1⟩·⟨1|0⟩ = {ip_factor:.0f}  (mixed-product property ✓)")
```

## Tensor Products

Tensor products are a way to combine vector spaces. One of the postulates of quantum mechanics is that the state of a qubit is completely described by a unit vector in a Hilbert space. To describe many qubits as a system, we need to combine all their Hilbert spaces into one bigger Hilbert space.

If we have a Hilbert space H and another Hilbert space J:

$$M = H \otimes J$$

If H is h-dimensional and J is j-dimensional:

$$\dim(M) = h \cdot j$$

### The tensor product of vectors

Notation (all equivalent):

$$|u\rangle \otimes |v\rangle = |u\rangle|v\rangle = |u, v\rangle = |uv\rangle$$

Properties:
- $s(|u\rangle \otimes |v\rangle) = (s|u\rangle) \otimes |v\rangle = |u\rangle \otimes (s|v\rangle)$
- Left distributive: $(|u\rangle + |z\rangle) \otimes |v\rangle = |u\rangle \otimes |v\rangle + |z\rangle \otimes |v\rangle$
- Right distributive: $|u\rangle \otimes (|v\rangle + |w\rangle) = |u\rangle \otimes |v\rangle + |u\rangle \otimes |w\rangle$
- **Not commutative**: $|u\rangle \otimes |v\rangle \neq |v\rangle \otimes |u\rangle$ in general

The **Kronecker product** is the implementation for arrays of numbers:

$$\begin{bmatrix} x_1 \\ x_2 \end{bmatrix} \otimes \begin{bmatrix} y_1 \\ y_2 \end{bmatrix} = \begin{bmatrix} x_1 y_1 \\ x_1 y_2 \\ x_2 y_1 \\ x_2 y_2 \end{bmatrix}$$

Example:

$$|01\rangle = |0\rangle \otimes |1\rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix} \otimes \begin{bmatrix} 0 \\ 1 \end{bmatrix} = \begin{bmatrix} 0 \\ 1 \\ 0 \\ 0 \end{bmatrix}$$

### The tensor product of operators

For operators A in V and B in U:

$$(A \otimes B)(|v\rangle \otimes |u\rangle) = A|v\rangle \otimes B|u\rangle$$

### The Kronecker product of matrices

Given an m × n matrix A and a p × q matrix B, their Kronecker product is an (m⋅p) × (n⋅q) matrix:

$$A \otimes B = \begin{bmatrix} a_{11}B & a_{12}B & \cdots \\ a_{21}B & a_{22}B & \cdots \\ \vdots & \vdots & \ddots \end{bmatrix}$$

### The inner product of composite vectors

For vectors |v⟩, |x⟩ in V and |u⟩, |z⟩ in U, with composite vectors |w⟩ = |v⟩⊗|u⟩ and |y⟩ = |x⟩⊗|z⟩:

$$\langle w|y\rangle = \langle v|x\rangle \cdot \langle u|z\rangle \quad \text{(11)}$$

This is due to the mixed-product property of the Kronecker product.
