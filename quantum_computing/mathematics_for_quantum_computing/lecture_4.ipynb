## Subspaces

Let's say you have a set, U, of vectors and it is a subset of a set, V, of vectors (U ⊆ V).

> *Figure 4.1 – The set U as a subset of V*

Is it possible that U is a subspace of V? Well, yes. It has met the first condition for subspaces, namely, that the potential subspace has to be a subset of the bigger vector space's set of vectors. What's next? Well, U also has to be a vector space using the same field as the vector space V. It has been proven that we only need to check for three small conditions.

### Definition

For a subset, U, of a vector space, V, with an associated field, F, of scalars, U is a subspace if:

1. The **0** vector is included in the subset U. (**0** ∈ U).
2. The subset U is closed for vector addition. If |u⟩ and |v⟩ ∈ U, then |u⟩ + |v⟩ ∈ U.
3. The subset U is closed for scalar multiplication. If |u⟩ ∈ U and s is a scalar, then s|u⟩ ∈ U.

> *Figure 4.2 – Ensuring closure for a potential subspace*

### Examples

Let's use two-dimensional real space, ℝ², as our overall vector space. What if we picked the line y = x on the X-Y plane to possibly be a subspace?

> *Figure 4.3 – The graph of y = x*

Let's call every vector on that line (for example, (1,1), (2,2), (-3, -3)) the set, W, of vectors:

$$W = \{(a, b) \in \mathbb{R}^2 \mid a = b\}$$

Let's check whether W is a subspace of ℝ²:

1. **Zero vector:** The **0** vector in ℝ² is (0,0), and that meets the condition. ✓
2. **Closure for addition:** If (a,a) + (b,b) = (a+b, a+b), which is still in W. ✓
3. **Closure for scalar multiplication:** s(a,a) = (sa, sa), which is still in W. ✓

So, W is a subspace of ℝ².

Now consider the line y = x + 1:

> *Figure 4.4 – The graph of y = x + 1*

Is this one a subspace of ℝ²? **No.** It fails the very first condition: it does not contain the zero vector.

**Exercise 1:** Test the following subsets of ℝ² and determine which ones are subspaces and which are not. Assume all variables are real numbers.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Subspace visualization ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Subspace: y = x (passes through origin)
ax = axes[0]
t = np.linspace(-4, 4, 100)
ax.plot(t, t, 'b-', linewidth=2, label='W: y = x (subspace)')
ax.quiver(0, 0, 1, 1, angles='xy', scale_units='xy', scale=1, color='green', linewidth=2, label='(1,1) ∈ W')
ax.quiver(0, 0, -2, -2, angles='xy', scale_units='xy', scale=1, color='purple', linewidth=2, label='(-2,-2) ∈ W')
ax.plot(0, 0, 'ko', markersize=8, label='0 ∈ W ✓')
ax.set_xlim(-4, 4); ax.set_ylim(-4, 4)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend(fontsize=9)
ax.set_title('W = {(a,b) ∈ ℝ² | a = b}  →  IS a subspace')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# Not a subspace: y = x + 1 (does NOT pass through origin)
ax = axes[1]
ax.plot(t, t + 1, 'r-', linewidth=2, label='y = x + 1 (NOT subspace)')
ax.plot(0, 0, 'ko', markersize=8)
ax.annotate('Origin (0,0)\nNOT on the line!', xy=(0, 0), xytext=(1.5, -2),
            fontsize=10, arrowprops=dict(arrowstyle='->', color='black'),
            bbox=dict(boxstyle='round', facecolor='lightyellow'))
ax.set_xlim(-4, 4); ax.set_ylim(-4, 4)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend(fontsize=9)
ax.set_title('y = x + 1  →  NOT a subspace (no zero vector)')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()

# Verify subspace conditions programmatically for W: y = x
def in_W(v):
    return np.isclose(v[0], v[1])

u = np.array([2, 2])
v = np.array([-3, -3])
s = 5.0
print("Subspace test for W = {(a,a) | a ∈ ℝ}:")
print(f"  Zero vector (0,0) ∈ W? {in_W([0,0])}")
print(f"  u={u}, v={v}, u+v={u+v}  ∈ W? {in_W(u+v)}")
print(f"  {s}·u = {s*u}  ∈ W? {in_W(s*u)}")
print("  → All conditions met: W is a subspace ✓")
```

## Linear Independence

### Linear combination

We learned in Chapter 2, Superposition with Euclid, that linear combinations are the scaling and addition of vectors. A more precise definition:

A linear combination for vectors |x₁⟩, |x₂⟩, … |xₙ⟩ and scalars c₁, c₂, … cₙ in a vector space, V, is a vector of the form:

$$|v\rangle = c_1|x_1\rangle + c_2|x_2\rangle + \cdots + c_n|x_n\rangle$$

### Linear dependence

If you have a set of vectors and you can create a linear combination of one of the vectors from a subset of the other vectors, then all of those vectors are linearly dependent.

Are |0⟩ and |1⟩ linearly dependent? No, because there are no scalars that you can multiply |0⟩ by to get |1⟩ and vice-versa.

A set of vectors is either linearly dependent or linearly independent! They cannot be both and they have to be one or the other. In fact:

> **A set of vectors is linearly independent if they are not linearly dependent.**

> *Figure 4.5 – Screenshot from Wolfram Alpha*

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Linear Independence ---
# Two vectors are linearly independent if neither is a scalar multiple of the other

# Independent vectors
v1 = np.array([1, 0])
v2 = np.array([0, 1])

# Dependent vectors (v4 = 2*v3)
v3 = np.array([1, 2])
v4 = np.array([2, 4])

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

ax = axes[0]
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', linewidth=2, label='|0⟩ = [1,0]')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', linewidth=2, label='|1⟩ = [0,1]')
ax.set_xlim(-0.5, 1.5); ax.set_ylim(-0.5, 1.5)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend()
ax.set_title('Linearly INDEPENDENT\n(cannot write one as scalar × other)')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

ax = axes[1]
ax.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='blue', linewidth=2, label='v₃ = [1,2]')
ax.quiver(0, 0, v4[0], v4[1], angles='xy', scale_units='xy', scale=1, color='red', linewidth=2, label='v₄ = [2,4] = 2·v₃')
ax.set_xlim(-0.5, 3); ax.set_ylim(-0.5, 5)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend()
ax.set_title('Linearly DEPENDENT\n(v₄ = 2·v₃)')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()

# Check linear independence using numpy (matrix rank)
M_indep = np.column_stack([v1, v2])
M_dep = np.column_stack([v3, v4])
print(f"Rank of [v1, v2] = {np.linalg.matrix_rank(M_indep)}  (= 2 → independent)")
print(f"Rank of [v3, v4] = {np.linalg.matrix_rank(M_dep)}  (< 2 → dependent)")
```

## Span

The span uses a set of distinct, indexed vectors to generate a vector space. How does it do this? It uses every possible linear combination of the set of vectors.

Let's start with a set, S, of vectors with just one vector:

> *Figure 4.6 – Graph of the one vector in S*

What would be its span? We can only scale this vector. If we scale it for all real numbers, it becomes a line!

> *Figure 4.7 – Graph of the span of S*

This line is the line y = x from the previous section. So, we have created a subspace of ℝ² with one vector using the span operator.

If we add a linearly dependent vector, it buys us nothing — span(S) = span(T).

To really get something, we need to add vectors that are linearly independent to the vectors already in our spanning set. If we add a linearly independent vector, we can fill the entire vector space of ℝ²!

$$\text{span}(U) = \mathbb{R}^2$$

Can we find a set with fewer vectors that spans ℝ²? It just so happens that we can! A set of linearly independent vectors. This condition of linear independence for a spanning set ends up being so important that we give spanning sets of linearly independent vectors a special name – a **basis**.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Span and Basis visualization ---
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Span of one vector = a line
ax = axes[0]
v = np.array([1, 1])
scalars = np.linspace(-3, 3, 50)
points = np.array([s * v for s in scalars])
ax.scatter(points[:, 0], points[:, 1], c=scalars, cmap='coolwarm', s=20, zorder=2)
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='black', linewidth=2, label='v = [1,1]')
ax.set_xlim(-4, 4); ax.set_ylim(-4, 4)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend()
ax.set_title('span({v}) = a line\n(1D subspace of ℝ²)')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# Span of two independent vectors = all of ℝ²
ax = axes[1]
e1 = np.array([1, 0])
e2 = np.array([0, 1])
# Show a grid of linear combinations
for a in np.linspace(-3, 3, 7):
    for b in np.linspace(-3, 3, 7):
        p = a * e1 + b * e2
        ax.plot(p[0], p[1], 'o', color='steelblue', markersize=3, alpha=0.6)
ax.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='blue', linewidth=2, label='|0⟩')
ax.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='red', linewidth=2, label='|1⟩')
ax.set_xlim(-4, 4); ax.set_ylim(-4, 4)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend()
ax.set_title('span({|0⟩, |1⟩}) = ℝ²\n(Standard/Computational basis)')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# Same vector in different bases
ax = axes[2]
u = np.array([3, 3])
# Standard basis
ax.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1, color='blue', linewidth=1.5, alpha=0.5, label='|0⟩')
ax.quiver(0, 0, 0, 1, angles='xy', scale_units='xy', scale=1, color='red', linewidth=1.5, alpha=0.5, label='|1⟩')
# Alternative basis
f1 = np.array([1, 1])
f2 = np.array([1, -1])
ax.quiver(0, 0, f1[0], f1[1], angles='xy', scale_units='xy', scale=1, color='green', linewidth=1.5, alpha=0.5, label='f₁ = [1,1]')
ax.quiver(0, 0, f2[0], f2[1], angles='xy', scale_units='xy', scale=1, color='orange', linewidth=1.5, alpha=0.5, label='f₂ = [1,-1]')
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='black', linewidth=2.5, label=f'|u⟩ = {u}')
ax.set_xlim(-1, 5); ax.set_ylim(-2, 5)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3); ax.legend(fontsize=8)
ax.set_title('Same vector, different bases\nStandard: (3,3)  Alt: (3,0)')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()

# Coordinates in different bases
F = np.column_stack([f1, f2])  # basis matrix
coords_F = np.linalg.solve(F, u)
print(f"|u⟩ = {u}")
print(f"Standard basis coords: ({u[0]}, {u[1]})")
print(f"Alt basis F = {{[1,1], [1,-1]}} coords: ({coords_F[0]}, {coords_F[1]})")
print(f"Verify: {coords_F[0]}*[1,1] + {coords_F[1]}*[1,-1] = {coords_F[0]*f1 + coords_F[1]*f2}")
```

## Basis

> **Basis**, ba·sis \ ˈbā-səs \ plural bases\ ˈbā-ˌsēz \ Noun
>
> Something on which something else is established or based.

You can choose different bases for a vector space. While the vector space itself does not change when you choose a different basis, the way things are described with numbers does.

Consider the vector |u⟩ with coordinates (3,3) in the standard basis.

> *Figure 4.11 – Graph of the vector |u⟩*

The **standard basis** (or **computational basis** in quantum computing) for ℝ² is:

$$|0\rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad |1\rangle = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$$

What's special about a basis is that there is a **unique** linear combination to describe each vector and the scalars used in this unique linear combination are called **coordinates**.

The unique linear combination to describe |u⟩ using the standard basis is:

$$|u\rangle = 3|0\rangle + 3|1\rangle$$

Now, consider a different basis F:

> *Figure 4.12 – Graph of the vectors in set F*

The coordinates of |u⟩ according to our basis F are (3,0). The vector is still the same — same length and direction — but with a new basis, we use different coordinates to describe it.

As Shakespeare wrote in Romeo and Juliet:

> *What's in a name? That which we call a rose*
> *By any other name would smell as sweet;*

Or, in other words:

> *What's in coordinates? That which we call a vector*
> *By any other coordinates would still be the same vector;*

If no basis is given, then we assume we are using the computational basis. It should be noted that the basis has to be indexed and ordered so that our coordinates do not get mixed up.

## Dimension

The dimension of a vector space ends up being very easy to define once you know what a basis is. One thing we didn't talk about is that the basis is the minimum set of vectors needed to span a space, and that the number of vectors in a basis for a particular vector space is always the same.

From this, we define the **dimension** of a vector space to be equal to the number of vectors it takes as a basis to describe a vector space. Equivalently, we could say that it is the number of coordinates it takes to describe a vector in the vector space. It follows that the dimension of ℝ² is two, ℝ³ is three, and ℝⁿ is n.
