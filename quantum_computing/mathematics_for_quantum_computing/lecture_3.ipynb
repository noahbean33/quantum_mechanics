## Sets

Sets are very intuitive and are really about grouping things together. For example, all mammals – taken together – form a set. In this set, its members are things such as the fox, squirrel, and dog. Sets don't care about duplication – so, if we have 5,000 dogs in our first mammals set, this set is equal to a set of mammals that has only one dog. Let's make this more formal.

### The definition of a set

A set is a collection of objects. This collection can be finite or infinite. Mathematical objects are abstract, have properties, and can be acted upon by operations. Examples of objects are numbers, functions, shapes, and matrices. Objects in a set are called elements or members.

### Notation

There are multiple ways to denote a set. The easiest way is to just describe it, as I did with the set of mammals.

### Set-builder notation

Set-builder notation is definitely more formal, but with this formality, you gain preciseness (which mathematicians covet). The easiest way to denote a set in set-builder notation is just to enumerate all the members of a set. You do this by giving the variable name of the set, followed by an equals sign, and then the members of the set are put into curly brackets and separated by commas.

An ellipsis (…) is used to skip listing elements if a pattern is clear or to denote an infinite set.

The final way you can denote a set is to include conditions for the members of your set.

> *Figure 3.1 – An annotated description of set-builder notation*

### Other set notation

An important symbol in set notation is ϵ, which denotes membership. If there is a slash through ϵ, it means the object is not a member of the set.

The next symbol to consider is used to denote subsets. If X and Y are sets, and every element of X is also an element of Y, then:
- X is a subset of Y, denoted by X ⊆ Y.
- Y is a superset of X, denoted by Y ⊇ X.

Finally, it is important to define the empty set that has no members and is denoted by ∅ or {}. The empty set is a subset of all sets.

### Important sets of numbers

Since mathematics is all about numbers, special attention should be given to certain sets of numbers that we will see in this book. Each one has a special double-struck capital letter to represent them:

- **ℕ** — the set of natural numbers defined as {0, 1, 2, 3, …}. This is the first set of numbers you learn as a child, and they are used for counting.
- **ℤ** — the set of integers defined as {…, -3, -2, -1, 0, 1, 2, 3, …}. This is a superset of ℕ. The letter Z comes from the German word Zahlen, which means numbers.
- **ℚ** — the set of rational numbers, where a rational number is defined as any number that can be expressed as a ratio or quotient of two integers. Since all integers are divisible by 1, ℤ is a subset of ℚ.
- **ℝ** — the set of real numbers. The real numbers are composed of ℚ and all the irrational numbers. Irrational numbers, when represented as a decimal, do not terminate, nor do they end with a repeating sequence. Some examples of irrational numbers: all square roots of natural numbers that are not perfect squares; the golden ratio, ϕ; Euler's number, e.
- **ℂ** — the set of complex numbers. A complex number is represented by: $z = a + bi$. If we set b = 0, then we have the set of all real numbers, so ℝ ⊆ ℂ.

To sum up:

$$\mathbb{N} \subseteq \mathbb{Z} \subseteq \mathbb{Q} \subseteq \mathbb{R} \subseteq \mathbb{C}$$

> *Figure 3.2 – A diagram showing all the sets of numbers*

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product as cart_product

# --- Sets in Python ---
A = {1, 2, 3}
B = {2, 3, 1}  # same as A (sets ignore order)
print(f"A = {A}")
print(f"B = {B}")
print(f"A == B? {A == B}  (sets ignore order and duplicates)\n")

# Set with duplicates collapses
C = {1, 1, 2, 2, 3}
print(f"{{1, 1, 2, 2, 3}} becomes {C}\n")

# Membership
print(f"2 ∈ A? {2 in A}")
print(f"5 ∈ A? {5 in A}\n")

# Subsets
N_sample = {0, 1, 2, 3, 4}         # ℕ (sample)
Z_sample = {-2, -1, 0, 1, 2, 3, 4} # ℤ (sample)
print(f"ℕ_sample ⊆ ℤ_sample? {N_sample.issubset(Z_sample)}")

# Cartesian product
X = {1, 2}
Y = {'a', 'b', 'c'}
cartesian = set(cart_product(X, Y))
print(f"\nX = {X}")
print(f"Y = {Y}")
print(f"X × Y = {cartesian}")
print(f"|X × Y| = {len(cartesian)}")

# Visualize nested number sets
fig, ax = plt.subplots(figsize=(8, 5))
from matplotlib.patches import Ellipse
sets_info = [
    ('ℂ (Complex)', 5.0, 3.2, 'lavender'),
    ('ℝ (Real)', 4.0, 2.6, 'lightyellow'),
    ('ℚ (Rational)', 3.0, 2.0, 'honeydew'),
    ('ℤ (Integer)', 2.0, 1.4, 'lightcyan'),
    ('ℕ (Natural)', 1.0, 0.8, 'mistyrose'),
]
for name, w, h, color in sets_info:
    ellipse = Ellipse((0, 0), w*2, h*2, facecolor=color, edgecolor='black', linewidth=1.5)
    ax.add_patch(ellipse)
    ax.text(w - 0.3, h - 0.2, name, fontsize=9, fontweight='bold', ha='center')

ax.set_xlim(-6, 6); ax.set_ylim(-4, 4)
ax.set_aspect('equal'); ax.set_title('Nested Number Sets: ℕ ⊆ ℤ ⊆ ℚ ⊆ ℝ ⊆ ℂ', fontsize=13)
ax.axis('off')
plt.tight_layout()
plt.show()
```

### Tuples

It is important to note that sets do not care about order. So, if set A = {1, 2, 3} and set B = {2, 3, 1}, A and B are equal. Sets also do not care about duplication. A mathematical object that does care about these things is called a tuple.

A tuple is a finite, ordered list of elements, which is denoted with open and close parentheses. Since order and duplication matters to tuples, different orderings are not equal. The number of elements in a tuple is defined as n, and we use this number to refer to a tuple as an n-tuple. A 2-tuple is also known as an ordered pair.

### The Cartesian product

The Cartesian product takes two sets and creates a third set of ordered pairs (that is, 2-tuples) from those two sets. It is denoted by the × symbol.

> *Figure 3.3 – An example of the Cartesian product of A × B*

It should be noted that the Cartesian product is not commutative, so in general, A × B ≠ B × A.

The greatest example of this operation is from Rene Descartes (who the Cartesian product is named after). The Cartesian plane is the Cartesian product of the set X of the real numbers ℝ and the set Y of the real numbers ℝ, denoted by ℝ × ℝ = ℝ².

> *Figure 3.4 – A Cartesian plane with Cartesian coordinates*

In general:

$$\mathbb{R}^n = \underbrace{\mathbb{R} \times \mathbb{R} \times \cdots \times \mathbb{R}}_{n \text{ times}}$$

So, for shorthand, we write ℝⁿ to denote all of the n-tuples of real numbers.

## Functions

Functions are fundamental to mathematics, and there is no doubt that you have been exposed to them before. However, I want to go over certain aspects of them in depth, as we will define things such as matrices as representations of functions later in the book.

### The definition of a function

A function, for example, y = f(x), maps every element x in a set A to another element y in set B. Each element y is called the image of x under the function f(x). Set A is called the domain of the function and set B is called the codomain of the function. The domain and codomain of a function are denoted by f: A → B.

> *Figure 3.5 – An example function*

All the images of f(x) form a set called the range. The range is a subset of the codomain.

There are two rules that functions must follow:
1. Every member of the domain must be mapped.
2. Every member of the domain cannot be mapped to more than one element in the codomain.

> *Figure 3.6 – An example of an illegal function*

**Exercise 1:** For the sets E = {a, b, c} and F = {4, 5, 6, 7, 8}, and a function, f:E → F, which of the following tables do not represent a function?

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Visualizing function types: injective, surjective, bijective ---
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

def draw_function(ax, title, domain, codomain, mapping, color='steelblue'):
    """Draw a function mapping diagram."""
    d_y = np.linspace(0, 1, len(domain))
    c_y = np.linspace(0, 1, len(codomain))
    for i, d in enumerate(domain):
        ax.plot(0, d_y[i], 'o', color='blue', markersize=10)
        ax.text(-0.15, d_y[i], str(d), fontsize=11, ha='center', va='center')
    for j, c in enumerate(codomain):
        ax.plot(1, c_y[j], 'o', color='red', markersize=10)
        ax.text(1.15, c_y[j], str(c), fontsize=11, ha='center', va='center')
    for d_val, c_val in mapping:
        d_idx = domain.index(d_val)
        c_idx = codomain.index(c_val)
        ax.annotate('', xy=(0.95, c_y[c_idx]), xytext=(0.05, d_y[d_idx]),
                     arrowprops=dict(arrowstyle='->', color=color, lw=1.5))
    ax.set_xlim(-0.4, 1.4); ax.set_ylim(-0.2, 1.2)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.text(0, -0.1, 'Domain', ha='center', fontsize=9)
    ax.text(1, -0.1, 'Codomain', ha='center', fontsize=9)
    ax.axis('off')

# Injective (one-to-one, but not onto)
draw_function(axes[0], 'Injective (not surjective)',
              ['a', 'b'], [1, 2, 3],
              [('a', 1), ('b', 3)])

# Surjective (onto, but not one-to-one)
draw_function(axes[1], 'Surjective (not injective)',
              ['a', 'b', 'c'], [1, 2],
              [('a', 1), ('b', 2), ('c', 1)], color='green')

# Bijective (both injective and surjective → invertible!)
draw_function(axes[2], 'Bijective (invertible!)',
              ['a', 'b', 'c'], [1, 2, 3],
              [('a', 2), ('b', 3), ('c', 1)], color='purple')

plt.tight_layout()
plt.show()

# Demonstrate f(x) = x² is not surjective over ℝ → ℝ
x = np.linspace(-3, 3, 300)
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, x**2, 'b-', linewidth=2, label='f(x) = x²')
ax.axhline(y=0, color='k', linewidth=0.5)
ax.fill_between(x, -5, 0, alpha=0.15, color='red', label='Negative reals: never reached!')
ax.set_xlim(-3, 3); ax.set_ylim(-2, 9)
ax.set_xlabel('x (domain: ℝ)'); ax.set_ylabel('f(x) (codomain: ℝ)')
ax.set_title('f(x) = x² is NOT surjective (ℝ → ℝ)')
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Invertible functions

Invertible functions are key in quantum computing because the laws of quantum mechanics only allow these types of functions in certain situations.

#### Injective functions

An injective function, also known as a one-to-one function, is a function where each element in the range is the image of only one element in the domain. It is important to note that not every element in the codomain needs to be in the range.

> *Figure 3.9 – The function on the left is injective and the function on the right is not*

#### Surjective functions

A surjective function, also known as an onto function, is a function where the range of the domain of the function is equal to its codomain. Another way to say this is that every element in the codomain is mapped to by at least one element in the domain.

> *Figure 3.10 – Two surjective functions*

The function f:ℝ →ℝ, where f(x) = x², is not surjective because not all of the elements in the codomain (namely, all negative real numbers) are mapped to.

#### Bijective functions

A bijective function is one that is both injective and surjective. And a function is invertible if it is bijective!

> *Figure 3.11 – A bijective, invertible function*

The inverse of a function, such as f(x), is usually denoted by f⁻¹(x). It makes sense that an invertible function has to be bijective. The only way I can make f⁻¹(x) a function is to make it follow the two rules for functions:
1. Every member of the domain must be mapped.
2. Every member of the domain cannot be mapped to more than one element in the codomain.

For f⁻¹(x), the domain and codomain are flipped. For it to follow the first rule, every element of the codomain for f(x) must be mapped (surjective). For the second rule, every member of the range of f(x) cannot be mapped to more than one element in its domain (injective).

## Binary Operations

You are probably familiar with some binary operations, for example, addition and multiplication, but we are going to look at binary operations in more depth.

### The definition of a binary operation

A binary operation is simply a function that takes two input values and outputs one value. More precisely, it takes an ordered pair (known as an operand) from the Cartesian product of two sets and produces an element in another set.

Within the number systems, addition is a good example of a binary operation. Let's define it for the real numbers:

$$+: \mathbb{R} \times \mathbb{R} \to \mathbb{R}$$

### Properties

Operations can have several properties. Here are the properties for a set S:

- **Identity:** There exists an identity element, e ∈ S, such that for all x ∈ S, a ֎ e = e ֎ a = a. This element, e, is unique, and it is called the identity element of the group.
- **Associativity:** If a, b, c ∈ S, then a ֎ (b ֎ c) = (a ֎ b) ֎ c.
- **Invertibility:** For every a ∈ S, there exists an a⁻¹, such that a ֎ a⁻¹ = a⁻¹ ֎ a = e, where e is the identity element.
- **Closure:** For every a, b ∈ A, a ֎ b produces an element c that is also in the set A. f: A × A → A.
- **Commutativity:** If a, b ∈ S, then a ֎ b = b ֎ a.

## Groups

A group builds upon the concept of a set by adding a binary operation to it. We denote a group by putting the set and the operation in angle brackets (⟨⟩). For example, ⟨A, ֎⟩ for set A and operation ֎. The operation has to follow certain rules to be considered a group, namely, the rules of **identity, associativity, invertibility, and closure**. If the operation ֎ also has the property of commutativity, then it is called an **Abelian group** (also known as a commutative group).

Now, let's look at a mathematical example. What if we define ֎ to be addition over the natural numbers ℕ denoted ⟨ℕ, ֎⟩ – is this a group? Let's go through the properties:

1. **Identity:** Does there exist an identity element e, such that a + e = e + a = a? Well, yes, if we define e = 0!
2. **Associativity:** Does a + (b + c) = (a + b) + c? Yes!
3. **Invertibility:** For every a ∈ ℕ, is there an a⁻¹, such that a + a⁻¹ = 0? Without negative numbers, there is no way to define an inverse. **Fails!**
4. **Closure:** If we take two numbers, a and b, in ℕ, then does a + b produce a natural number? Yes!

So, addition with the set ℕ is **not** a group. Is there a set that would work? Yes! The set of all integers, ℤ! We can then define a⁻¹ to be –a, and invertibility is fulfilled because a + (-a) = 0! Therefore, ⟨ℤ, +⟩ qualifies as a group! Since addition is commutative as well, the group is also an Abelian group.

## Fields

Fields extend the concept of groups to include another operation. A field is a set (denoted by S) and two operations (+ and ⋅) that we will notate as {S, +, ⋅}, which follows these rules:

1. ⟨S, +⟩ is an Abelian group with the identity element e = 0.
2. If you exclude the number 0 from the set S to produce a new set S', then ⟨S', ⋅ ⟩ is an Abelian group with the identity element e = 1.
3. For the rule of distributivity, let a, b, c ∈ S. Then, a ⋅ (b + c) = a ⋅ b + a ⋅ c.

The set of real numbers ℝ, with the operations of addition and multiplication, is the most obvious example of a field, but there are plenty of others.

**Exercise 2:** Is the set ℝ, with the operations of subtraction and division {ℝ, −, ÷}, a field?

```python
import numpy as np

# --- Verifying vector space axioms numerically ---
# Vector space: ℝ² over the field ℝ

u = np.array([3, -1])
v = np.array([2, 5])
zero = np.array([0, 0])
alpha, beta = 2.5, -3.0

print("=== Abelian Group ⟨ℝ², +⟩ ===")
print(f"|u⟩ = {u},  |v⟩ = {v}\n")

# Identity
print(f"Identity:       |u⟩ + 0 = {u + zero}  == |u⟩? {np.array_equal(u + zero, u)}")

# Associativity
w = np.array([-1, 4])
print(f"Associativity:  (u+v)+w = {(u+v)+w},  u+(v+w) = {u+(v+w)},  equal? {np.array_equal((u+v)+w, u+(v+w))}")

# Invertibility
print(f"Invertibility:  |u⟩ + (-|u⟩) = {u + (-u)}  == 0? {np.array_equal(u + (-u), zero)}")

# Closure (sum of two real vectors is still a real vector)
print(f"Closure:        |u⟩ + |v⟩ = {u + v}  (still in ℝ²)")

# Commutativity
print(f"Commutativity:  u+v = {u+v},  v+u = {v+u},  equal? {np.array_equal(u+v, v+u)}")

print("\n=== Scalar Multiplication Axioms ===")
print(f"α = {alpha},  β = {beta}\n")

# Compatibility: α(β|v⟩) = (αβ)|v⟩
lhs = alpha * (beta * v)
rhs = (alpha * beta) * v
print(f"Compatibility:      α(β|v⟩) = {lhs},  (αβ)|v⟩ = {rhs},  equal? {np.allclose(lhs, rhs)}")

# Distributivity wrt vector addition
lhs = alpha * (u + v)
rhs = alpha * u + alpha * v
print(f"Distrib (vec add):  α(u+v) = {lhs},  αu+αv = {rhs},  equal? {np.allclose(lhs, rhs)}")

# Distributivity wrt field addition
lhs = (alpha + beta) * v
rhs = alpha * v + beta * v
print(f"Distrib (field add): (α+β)v = {lhs},  αv+βv = {rhs},  equal? {np.allclose(lhs, rhs)}")

# Multiplicative identity
print(f"Identity (×1):      1·|v⟩ = {1*v}  == |v⟩? {np.array_equal(1*v, v)}")
```

## Vector Space

Now that we have covered all of the abstract concepts we need to understand, we can give a formal definition of a vector space.

A vector space is defined as having the following mathematical objects:

1. An **Abelian group** ⟨V,+⟩ with an identity element e. We call members of the set V vectors. We define the identity element to be the zero vector, denoted by **0**. The operation + is called vector addition.
2. A **field** {F, +, ⋅}. We say that V is a vector space over the field F, and we call the members of F scalars.

> **Note:** The zero vector is denoted with a bold zero – **0** – and it is totally different from the vector |0⟩. This is the convention in quantum computing.

We can define an additional operation as **scalar multiplication**, which is an operation between a scalar and a vector:

- Let a scalar s ∈ F and the vector |v⟩ ∈ V. Scalar multiplication is a binary operation, f: F × V → V. The multiplicative identity element of the field F of scalars is 1: |v⟩ ⋅ 1 = 1 ⋅ |v⟩ = |v⟩.

This new operation must also be compatible with addition and multiplication from our field F of scalars and with vector addition. Let α, β ∈ F and |u⟩, |v⟩ ∈ V:

- **Compatibility with field multiplication:** α(β|v⟩) = (αβ)|v⟩
- **Distributivity with respect to vector addition:** α(|u⟩ + |v⟩) = α|u⟩ + α|v⟩
- **Distributivity with respect to field addition:** (α + β)|v⟩ = α|v⟩ + β|v⟩

In quantum computing, our vector spaces are concerned with the field of complex numbers, ℂ. However, for the next few chapters, we will stick to the field of real numbers, ℝ.
