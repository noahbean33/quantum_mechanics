## Measuring Angles

There are two main ways to measure angles – degrees and radians. Radians are almost exclusively used in quantum computing.

### Degrees

There are 360 degrees in a circle.

> *Figure 12.1 – Compass*

On the X-Y plane, zero degrees starts on the right side of the x axis. The X-Y plane can be broken up into four quadrants with each taking up 90 degrees.

> *Figure 12.2 – Four quadrants*

When degrees or radians are positive, you go counterclockwise; when negative, clockwise.

> *Figure 12.3 – Positive and negative angles*

### Radians

Radians are the most used measure in mathematics and physics because they make calculations much easier. The key conversion:

$$360° = 2\pi \text{ rad}$$

Conversion formulas:

$$\text{radians} = \text{degrees} \times \frac{\pi}{180}$$

$$\text{degrees} = \text{radians} \times \frac{180}{\pi}$$

> *Figure 12.5 – Degrees and equivalent radians on the X-Y plane*

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Degree/Radian Conversion ---
degrees = [0, 30, 45, 60, 90, 120, 135, 150, 180, 270, 360]
radians = [np.radians(d) for d in degrees]
print("=== Degree ↔ Radian Conversion ===")
for d, r in zip(degrees, radians):
    print(f"  {d:>3}° = {r:.4f} rad = {r/np.pi:.4f}π")

# --- Unit Circle ---
fig, ax = plt.subplots(figsize=(7, 7))
theta = np.linspace(0, 2*np.pi, 300)
ax.plot(np.cos(theta), np.sin(theta), 'k-', linewidth=1.5)

# Mark special angles
special_angles = {
    '0': 0, 'π/6': np.pi/6, 'π/4': np.pi/4, 'π/3': np.pi/3,
    'π/2': np.pi/2, '2π/3': 2*np.pi/3, '3π/4': 3*np.pi/4,
    '5π/6': 5*np.pi/6, 'π': np.pi, '7π/6': 7*np.pi/6,
    '5π/4': 5*np.pi/4, '4π/3': 4*np.pi/3, '3π/2': 3*np.pi/2,
    '5π/3': 5*np.pi/3, '7π/4': 7*np.pi/4, '11π/6': 11*np.pi/6,
}
for label, ang in special_angles.items():
    x, y = np.cos(ang), np.sin(ang)
    ax.plot(x, y, 'ro', markersize=5)
    ax.annotate(label, (x, y), textcoords="offset points",
                xytext=(12*np.sign(x) if abs(x) > 0.1 else 0,
                        12*np.sign(y) if abs(y) > 0.1 else 12),
                fontsize=7, ha='center')

ax.axhline(y=0, color='gray', linewidth=0.5)
ax.axvline(x=0, color='gray', linewidth=0.5)
ax.set_aspect('equal'); ax.set_xlim(-1.4, 1.4); ax.set_ylim(-1.4, 1.4)
ax.set_title('Unit Circle with Special Angles', fontsize=13)
ax.grid(True, alpha=0.2)

# Label quadrants
ax.text(0.5, 0.5, 'Q I\nsin+, cos+', ha='center', fontsize=9, color='green')
ax.text(-0.5, 0.5, 'Q II\nsin+, cos−', ha='center', fontsize=9, color='blue')
ax.text(-0.5, -0.5, 'Q III\nsin−, cos−', ha='center', fontsize=9, color='red')
ax.text(0.5, -0.5, 'Q IV\nsin−, cos+', ha='center', fontsize=9, color='orange')

plt.tight_layout()
plt.show()
```

## Trigonometric Functions

There are three main trigonometric functions, all based on right triangles.

> *Figure 12.6 – A right triangle*

The three trig functions (with angle θ at vertex A, right angle at C):

$$\sin\theta = \frac{\text{opposite}}{\text{hypotenuse}}, \quad \cos\theta = \frac{\text{adjacent}}{\text{hypotenuse}}, \quad \tan\theta = \frac{\text{opposite}}{\text{adjacent}}$$

Their inverses:

$$\csc\theta = \frac{1}{\sin\theta}, \quad \sec\theta = \frac{1}{\cos\theta}, \quad \cot\theta = \frac{1}{\tan\theta}$$

> *Figure 12.7 – The common values for sin, cos, and tan*

As angles get bigger, they move from the first quadrant into the other quadrants and become positive or negative:

> *Figure 12.8 – The trig functions by quadrant*

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Trig Functions: sin, cos, tan ---
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

x = np.linspace(-2*np.pi, 2*np.pi, 500)

# sin and cos
ax = axes[0, 0]
ax.plot(x, np.sin(x), 'b-', linewidth=2, label='sin(θ)')
ax.plot(x, np.cos(x), 'r-', linewidth=2, label='cos(θ)')
ax.set_title('sin(θ) and cos(θ)')
ax.legend(); ax.grid(True, alpha=0.3)
ax.set_xlim(-2*np.pi, 2*np.pi); ax.set_ylim(-1.5, 1.5)
ax.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax.set_xticklabels(['-2π', '-π', '0', 'π', '2π'])
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# tan
ax = axes[0, 1]
# Avoid asymptotes
x_tan = np.linspace(-1.5, 1.5, 300)
ax.plot(x_tan, np.tan(x_tan), 'g-', linewidth=2, label='tan(θ)')
ax.set_title('tan(θ) = sin(θ)/cos(θ)')
ax.legend(); ax.grid(True, alpha=0.3)
ax.set_ylim(-5, 5)
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# Phase relationship: cos(θ) = sin(θ + π/2)
ax = axes[1, 0]
ax.plot(x, np.sin(x), 'b-', linewidth=2, label='sin(θ)')
ax.plot(x, np.sin(x + np.pi/2), 'r--', linewidth=2, label='sin(θ + π/2) = cos(θ)')
ax.set_title('Phase Relationship: cos(θ) = sin(θ + π/2)')
ax.legend(); ax.grid(True, alpha=0.3)
ax.set_xlim(-2*np.pi, 2*np.pi); ax.set_ylim(-1.5, 1.5)
ax.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax.set_xticklabels(['-2π', '-π', '0', 'π', '2π'])
ax.axhline(y=0, color='k', linewidth=0.5)

# Pythagorean identity: sin²(θ) + cos²(θ) = 1
ax = axes[1, 1]
ax.plot(x, np.sin(x)**2, 'b-', linewidth=1.5, label='sin²(θ)', alpha=0.7)
ax.plot(x, np.cos(x)**2, 'r-', linewidth=1.5, label='cos²(θ)', alpha=0.7)
ax.plot(x, np.sin(x)**2 + np.cos(x)**2, 'k--', linewidth=2, label='sin²+cos² = 1')
ax.set_title('Pythagorean Identity: sin²(θ) + cos²(θ) = 1')
ax.legend(); ax.grid(True, alpha=0.3)
ax.set_xlim(-2*np.pi, 2*np.pi); ax.set_ylim(-0.2, 1.5)
ax.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax.set_xticklabels(['-2π', '-π', '0', 'π', '2π'])

plt.tight_layout()
plt.show()

# Common trig values table
print("=== Common Trig Values ===")
print(f"{'θ':>8} {'sin(θ)':>10} {'cos(θ)':>10} {'tan(θ)':>10}")
print("-" * 42)
for deg in [0, 30, 45, 60, 90]:
    rad = np.radians(deg)
    s = np.sin(rad)
    c = np.cos(rad)
    t = np.tan(rad) if deg != 90 else float('inf')
    print(f"{deg:>5}° {s:>10.4f} {c:>10.4f} {str(round(t, 4)):>10}")
```

## Formulas

### Tangent identity

$$\tan\theta = \frac{\sin\theta}{\cos\theta}$$

### Phase relationship

Sine and cosine are "out of phase" by π/2:

$$\cos\theta = \sin\left(\theta + \frac{\pi}{2}\right)$$

> *Figure 12.9 – The graphs of sine and cosine*

### Pythagorean identity

$$\sin^2\theta + \cos^2\theta = 1$$

## The Trig Cheat Sheet

### Pythagorean identities

$$\sin^2\theta + \cos^2\theta = 1$$
$$1 + \tan^2\theta = \sec^2\theta$$
$$1 + \cot^2\theta = \csc^2\theta$$

### Double angle identities

$$\sin(2\theta) = 2\sin\theta\cos\theta$$
$$\cos(2\theta) = \cos^2\theta - \sin^2\theta$$

### Sum/difference identities

$$\sin(\alpha \pm \beta) = \sin\alpha\cos\beta \pm \cos\alpha\sin\beta$$
$$\cos(\alpha \pm \beta) = \cos\alpha\cos\beta \mp \sin\alpha\sin\beta$$

### Product-to-sum identities

$$\sin\alpha\sin\beta = \frac{1}{2}[\cos(\alpha - \beta) - \cos(\alpha + \beta)]$$
$$\cos\alpha\cos\beta = \frac{1}{2}[\cos(\alpha - \beta) + \cos(\alpha + \beta)]$$
$$\sin\alpha\cos\beta = \frac{1}{2}[\sin(\alpha + \beta) + \sin(\alpha - \beta)]$$
