## The Inverse of a Matrix

It would be nice to have a way to do algebra on matrices the way we do for simple algebraic expressions. The inverse of a matrix provides us with a way to do this. Similar to the reciprocal for rational numbers:

$$a \cdot a^{-1} = 1$$

The inverse of a matrix is defined to be:

$$AA^{-1} = A^{-1}A = I$$

The matrix inverse can be used in algebraic manipulation. To find |x⟩ in A|x⟩ = |b⟩:

$$|x\rangle = A^{-1}|b\rangle$$

Remember: matrix multiplication is not commutative, so if you left multiply on one side, you must left multiply on the other side. Not all matrices have an inverse — determinants will help us here.

## Determinants

Determinants determine whether a square matrix is invertible. The determinant is a function from ℂⁿˣⁿ to ℂ — it takes an n × n square matrix and spits out a scalar.

For a 2 × 2 matrix:

$$\det\begin{pmatrix} a & b \\ c & d \end{pmatrix} = ad - bc$$

For a 3 × 3 matrix, the formula is considerably more involved. For larger matrices, use a calculator like WolframAlpha.

> *Figure 7.1 – The WolframAlpha online determinant calculator*

> **Determinants and Invertibility:** If the determinant of a matrix equals zero, it is not invertible. Otherwise, the matrix is invertible.

If a matrix is not invertible, we call it **singular** or **degenerate**.

**Exercise one:** What is the determinant of the given 2 × 2 matrices?

```python
import numpy as np

# --- Matrix Inverse and Determinant ---
A = np.array([[4, 7], [2, 6]])
det_A = np.linalg.det(A)
A_inv = np.linalg.inv(A)

print("=== Determinant & Inverse ===")
print(f"A =\n{A}\n")
print(f"det(A) = {det_A:.1f}")
print(f"A is invertible? {not np.isclose(det_A, 0)}\n")
print(f"A⁻¹ =\n{np.round(A_inv, 4)}\n")
print(f"A · A⁻¹ =\n{np.round(A @ A_inv, 10)}\n")

# 2x2 inverse formula: (1/det) * [[d, -b], [-c, a]]
a, b, c, d = A[0,0], A[0,1], A[1,0], A[1,1]
A_inv_manual = (1/det_A) * np.array([[d, -b], [-c, a]])
print(f"Manual 2×2 formula: (1/{det_A:.0f}) * [[{d}, {-b}], [{-c}, {a}]]")
print(f"= \n{np.round(A_inv_manual, 4)}")
print(f"Matches numpy? {np.allclose(A_inv, A_inv_manual)}\n")

# Singular matrix (det = 0)
B = np.array([[1, 2], [2, 4]])
print(f"B =\n{B}")
print(f"det(B) = {np.linalg.det(B):.1f}  → SINGULAR (not invertible)")

# Solve A|x⟩ = |b⟩ using the inverse
b_vec = np.array([1, 2])
x = A_inv @ b_vec
print(f"\nSolving A|x⟩ = |b⟩  where |b⟩ = {b_vec}")
print(f"|x⟩ = A⁻¹|b⟩ = {np.round(x, 4)}")
print(f"Verify: A|x⟩ = {np.round(A @ x, 4)}")
```

## The Invertible Matrix Theorem

The invertible matrix theorem is a great result in linear algebra. Let A be a square n × n matrix. If A is invertible (det(A) ≠ 0), then the following properties follow:

1. The column vectors of A are linearly independent.
2. The column vectors of A form a basis for ℂⁿ.
3. The row vectors of A form a basis for ℂⁿ and are also linearly independent.
4. The linear transformation mapping |x⟩ to A|x⟩ is a bijection from ℂⁿ to ℂⁿ.

If A is **not** invertible (det(A) = 0), then all the preceding properties are false.

## Calculating the Inverse of a Matrix

We can use the determinant to calculate the inverse. For a 2 × 2 matrix:

$$A^{-1} = \frac{1}{\det(A)}\begin{pmatrix} d & -b \\ -c & a \end{pmatrix}$$

The reciprocal of the determinant is used. If the determinant is zero, the formula is undefined — but the matrix is not invertible anyway.

Also note: a and d switch position, and b and c are multiplied by -1.

## Eigenvalues and Eigenvectors

The special vectors called eigenvectors are the ones that only get multiplied by a scalar when they are transformed. The scalar is called the **eigenvalue**.

> *Figure 7.2 – Reflection transformation*

> *Figure 7.3 – Vectors and their reflections*

For reflection:
- Any vector **perpendicular** to the axis of reflection has eigenvalue **-1**
- Any vector **parallel** to the axis of reflection has eigenvalue **1**

### Definition

Given a linear operator Â, an eigenvector is a non-zero vector that when transformed by Â is equal to being multiplied by a scalar λ (pronounced lamb-da):

$$\hat{A}|x\rangle = \lambda|x\rangle$$

This is also true for any matrix A that represents the linear operator:

$$A|x\rangle = \lambda|x\rangle$$

> *"What is a baby eigensheep?" "A lamb, duh!"*

### Eigenspace

An eigenvalue with its set of eigenvectors, plus the zero vector, constitutes a subspace called the **eigenspace**. The basis set of eigenvectors is called the **eigenbasis**.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Eigenvalues and Eigenvectors ---
# Reflection matrix over x-axis
Reflect = np.array([[1, 0], [0, -1]])
eigenvalues, eigenvectors = np.linalg.eig(Reflect)

print("=== Reflection over x-axis ===")
print(f"Matrix =\n{Reflect}\n")
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    vec = eigenvectors[:, i]
    print(f"λ_{i+1} = {lam:.1f},  eigenvector = {vec}")
    print(f"  Verify: A·v = {Reflect @ vec},  λ·v = {lam * vec}")

# More interesting example
A = np.array([[2, 1], [1, 2]])
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"\n=== Matrix A ===")
print(f"A =\n{A}\n")
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    vec = eigenvectors[:, i]
    print(f"λ_{i+1} = {lam:.4f},  eigenvector = {np.round(vec, 4)}")

# Visualize eigenvectors: they only get scaled, not rotated!
fig, ax = plt.subplots(figsize=(7, 7))

# Plot transformation of many vectors
n_arrows = 20
angles = np.linspace(0, 2*np.pi, n_arrows, endpoint=False)
for ang in angles:
    v = np.array([np.cos(ang), np.sin(ang)])
    Av = A @ v
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='lightblue', alpha=0.5, linewidth=1)
    ax.quiver(0, 0, Av[0], Av[1], angles='xy', scale_units='xy', scale=1, color='lightsalmon', alpha=0.5, linewidth=1)

# Highlight eigenvectors
for i in range(len(eigenvalues)):
    vec = eigenvectors[:, i]
    Avec = A @ vec
    ax.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color='blue', linewidth=3,
              label=f'eigenvec (λ={eigenvalues[i]:.1f})')
    ax.quiver(0, 0, Avec[0], Avec[1], angles='xy', scale_units='xy', scale=1, color='red', linewidth=3,
              linestyle='dashed')

ax.set_xlim(-3.5, 3.5); ax.set_ylim(-3.5, 3.5)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
ax.legend(fontsize=9)
ax.set_title('Eigenvectors (blue) only get scaled → (red)\nAll other vectors change direction')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)
plt.tight_layout()
plt.show()
```

### The characteristic equation

Starting from $A|x\rangle = \lambda|x\rangle$, we manipulate algebraically:

$$A|x\rangle - \lambda|x\rangle = \mathbf{0}$$
$$(A - \lambda I)|x\rangle = \mathbf{0} \quad \text{(2)}$$

For Equation (2) to have a non-trivial solution, the determinant must equal zero:

$$\det(A - \lambda I) = 0 \quad \text{(3)}$$

This is the **characteristic equation** of matrix A. The polynomial found is called the **characteristic polynomial**.

For larger matrices, use calculators like Symbolab (https://www.symbolab.com/).

## Trace

The trace of a matrix is the sum of the values on the main diagonal:

$$\text{tr}(A) = \sum_{i=1}^{n} a_{ii}$$

The trace of a linear operator is the same for any matrix that represents it.

```python
import numpy as np

# --- Trace and Special Properties of Eigenvalues ---
A = np.array([[2, 1], [1, 2]])
eigenvalues = np.linalg.eigvals(A)
trace_A = np.trace(A)
det_A = np.linalg.det(A)

print("=== Trace & Eigenvalue Properties ===")
print(f"A =\n{A}\n")
print(f"Trace(A) = {trace_A}")
print(f"det(A)   = {det_A:.4f}\n")
print(f"Eigenvalues: {eigenvalues}")
print(f"Sum of eigenvalues:     {np.sum(eigenvalues):.4f}  == Trace? {np.isclose(np.sum(eigenvalues), trace_A)}")
print(f"Product of eigenvalues: {np.prod(eigenvalues):.4f}  == Det?   {np.isclose(np.prod(eigenvalues), det_A)}")

# Pauli matrices (important in quantum computing)
print("\n=== Pauli Matrices ===")
X = np.array([[0, 1], [1, 0]])
Y = np.array([[0, -1j], [1j, 0]])
Z = np.array([[1, 0], [0, -1]])

for name, M in [('X', X), ('Y', Y), ('Z', Z)]:
    evals = np.linalg.eigvals(M)
    print(f"Pauli-{name}: eigenvalues = {np.round(evals, 4)}, trace = {np.trace(M):.1f}, det = {np.linalg.det(M):.1f}")
```

## The Special Properties of Eigenvalues

The sum of the eigenvalues equals the trace:

$$\sum_i \lambda_i = \text{tr}(A)$$

The product of all eigenvalues equals the determinant:

$$\prod_i \lambda_i = \det(A)$$
