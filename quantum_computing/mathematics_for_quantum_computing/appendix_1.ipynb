# Appendix 1: Bra–ket Notation

Bra-ket notation was introduced by Paul Dirac in 1939 and is sometimes called Dirac notation consequently. Kets are denoted by a pipe ("|") and right-angle bracket ("⟩"), like so – |Label⟩, while bras are denoted by a left-angle bracket ("⟨") and pipe ("|"), like so ⟨Label|. Kets represent vectors in a Hilbert space and bras represent their covectors in a dual Hilbert space. The labels for kets and bras can be lowercase letters, numbers, and greek letters. Uppercase letters are usually reserved for operators.

The computational basis vectors are represented by |0⟩ and |1⟩. It is important to note that the zero vector is denoted by **0** and is totally different than |0⟩. The zero vector, sometimes referred to as the null vector, is the only one that is not represented as a ket.

An inner product between two kets, |ϕ⟩ and |ψ⟩, is notated this way – ⟨ϕ|ψ⟩. This can be called a "bracket" and brings the notation full circle.

Other notations used are shown in the original reference table.

## Bras

A bra is a **linear functional**. These are a special case of linear transformation that takes in a vector and spits out a scalar:

$$f: V \to F$$

For instance, one could define a linear functional for every vector in ℝ²:

$$f_{\langle a|}(|x\rangle) = a_1 x_1 + a_2 x_2$$

The set of all linear functionals that can be defined on a vector space actually form their own vector space called the **dual vector space**.

Instead of using the usual function notation, Paul Dirac came up with a notation called a **bra**:

$$\langle v|$$

Since every vector has its own linear functional (called its dual vector or covector), the label between the angle bracket and vertical bar is the dual vector for the ket with the same label. In other words, every ket |v⟩ has a linear functional ⟨v| defined for it.

The function defined for each ket is the **inner product**:

$$\langle v|w\rangle = \sum_i v_i^* w_i$$

```python
import numpy as np

# --- Bra-ket Notation in NumPy ---
# Kets as column vectors
ket_0 = np.array([[1], [0]])
ket_1 = np.array([[0], [1]])

# Bras as conjugate transpose of kets
bra_0 = ket_0.conj().T
bra_1 = ket_1.conj().T

print("=== Computational Basis ===")
print(f"|0⟩ =\n{ket_0}\n")
print(f"|1⟩ =\n{ket_1}\n")
print(f"⟨0| = (|0⟩)† = {bra_0}")
print(f"⟨1| = (|1⟩)† = {bra_1}\n")

# Bracket (inner product) as matrix multiplication
print("=== Brackets (Inner Products) ===")
print(f"⟨0|0⟩ = {(bra_0 @ ket_0)[0,0]}  (normalized)")
print(f"⟨1|1⟩ = {(bra_1 @ ket_1)[0,0]}  (normalized)")
print(f"⟨0|1⟩ = {(bra_0 @ ket_1)[0,0]}  (orthogonal)")
print(f"⟨1|0⟩ = {(bra_1 @ ket_0)[0,0]}  (orthogonal)\n")

# Complex ket example
psi = np.array([[1+1j], [2-3j]])
bra_psi = psi.conj().T
print("=== Complex Vector Example ===")
print(f"|ψ⟩ =\n{psi}\n")
print(f"⟨ψ| = {bra_psi}")
print(f"⟨ψ|ψ⟩ = {(bra_psi @ psi)[0,0]}  (real, positive)")
print(f"||ψ|| = √⟨ψ|ψ⟩ = {np.sqrt((bra_psi @ psi)[0,0]).real:.4f}\n")

# Bra as a linear functional: ⟨v| maps any |w⟩ to a scalar
v = np.array([[1], [2]])
w = np.array([[3], [4]])
bra_v = v.conj().T
print("=== Bra as Linear Functional ===")
print(f"⟨v| = {bra_v}")
print(f"|w⟩ =\n{w}")
print(f"⟨v|w⟩ = {(bra_v @ w)[0,0]}  (maps vector → scalar)")
```
